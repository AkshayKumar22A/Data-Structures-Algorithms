# Deque (Double-Ended Queue) in Python

## What is a Deque?

A **Deque** (Double-Ended Queue) is a linear data structure that allows insertion and deletion of elements from **both ends**—front and rear. Unlike a regular queue (FIFO) or stack (LIFO), a deque combines the benefits of both, making it highly flexible.

### Visual Representation

```
Front                          Rear
  ↓                             ↓
[ 10 ]  ←→  [ 20 ]  ←→  [ 30 ]  ←→  [ 40 ]
  ↑                             ↑
Insert/Delete              Insert/Delete
from front                from rear (back)
```

---

## Common Methods

The `collections.deque` module in Python provides the following essential methods:

### 1. **Construction & Basic Operations**

```python
from collections import deque

# Create an empty deque
dq = deque()

# Create a deque from an iterable
dq = deque([1, 2, 3, 4, 5])

# Maximum size (optional; 0 = unlimited)
dq = deque(maxlen=5)
```

| Method | Time Complexity | Description |
|--------|-----------------|-------------|
| `len(dq)` | O(1) | Get the number of elements |
| `bool(dq)` | O(1) | Check if deque is empty |

### 2. **Rear (Right) End Operations**

```python
# Add element to the rear (back)
dq.append(x)              # O(1)

# Remove and return element from the rear
last = dq.pop()           # O(1)

# Peek at the rear without removing
rear_elem = dq[-1]        # O(1)
```

**Example:**
```python
dq = deque([10, 20, 30])
dq.append(40)             # dq = [10, 20, 30, 40]
val = dq.pop()            # val = 40; dq = [10, 20, 30]
peek = dq[-1]             # peek = 30 (no removal)
```

### 3. **Front (Left) End Operations**

```python
# Add element to the front
dq.appendleft(x)          # O(1)

# Remove and return element from the front
first = dq.popleft()      # O(1)

# Peek at the front without removing
front_elem = dq[0]        # O(1)
```

**Example:**
```python
dq = deque([20, 30, 40])
dq.appendleft(10)         # dq = [10, 20, 30, 40]
val = dq.popleft()        # val = 10; dq = [20, 30, 40]
peek = dq[0]              # peek = 20 (no removal)
```

### 4. **Rotation & Reversal**

```python
# Rotate elements to the right by n steps
dq.rotate(n)              # O(n)

# Reverse all elements in-place
dq.reverse()              # O(n)
```

**Example:**
```python
dq = deque([1, 2, 3, 4])
dq.rotate(1)              # dq = [4, 1, 2, 3] (right rotation)
dq.rotate(-1)             # dq = [1, 2, 3, 4] (left rotation)

dq.reverse()              # dq = [4, 3, 2, 1]
```

### 5. **Bulk Operations**

```python
# Extend from the right with multiple elements
dq.extend([x1, x2, x3])   # O(k) where k is the number of elements

# Extend from the left (inserts in reverse order)
dq.extendleft([x1, x2, x3])  # O(k)

# Clear all elements
dq.clear()                # O(n)
```

**Example:**
```python
dq = deque([10, 20])
dq.extend([30, 40, 50])   # dq = [10, 20, 30, 40, 50]

dq2 = deque([10])
dq2.extendleft([5, 4, 3]) # dq2 = [3, 4, 5, 10] (inserts in reverse)
```

### 6. **Count & Access**

```python
# Count occurrences of an element
count = dq.count(x)       # O(n)

# Check if element exists (in operator)
if x in dq:               # O(n)
    print("Found")

# Access by index (random access)
elem = dq[i]              # O(1) amortized
```

## ⚠️ Pitfalls to Avoid When Using Deque

### 1. **Confusing `appendleft()` with prepend**
```python
dq = deque([2, 3, 4])
dq.appendleft(1)          # ✓ Correct: dq = [1, 2, 3, 4]

# ✗ Wrong: This doesn't work as expected
dq.insert(0, 0)           # O(n) - inefficient for deque
```
**Lesson:** use `appendleft()` for O(1) insertion at front.

### 2. **Not handling `maxlen`**
```python
dq = deque(maxlen=3)
dq.extend([1, 2, 3, 4])   # dq = [2, 3, 4] - silently discards 1!
```
**Lesson:** Be aware that when `maxlen` is set, adding beyond capacity **silently removes** from the opposite end.

### 3. **Misunderstanding `extendleft()`**
```python
dq = deque([10])
dq.extendleft([1, 2, 3])  # dq = [3, 2, 1, 10] ← REVERSED ORDER!
```
**Lesson:** `extendleft()` reverses the input order because it calls `appendleft()` sequentially.

### 4. **Using regular indexing for performance-critical code**
```python
# ✗ Inefficient - O(n) per access in worst case
for i in range(len(dq)):
    print(dq[i])

# ✓ Efficient - O(1) iteration
for elem in dq:
    print(elem)
```

### 5. **Forgetting thread safety**
```python
# ✗ Not thread-safe
dq.append(1)
if len(dq) > 0:
    dq.pop()  # Could be empty by here in multi-threaded code!

# ✓ For thread-safe operations, use queue.Queue instead
```

## 🎯 Real-World Applications of Deque

### 1. **Implementing Stacks & Queues**
```python
# Stack (LIFO) using deque
class Stack:
    def __init__(self):
        self.dq = deque()
    
    def push(self, x):
        self.dq.append(x)
    
    def pop(self):
        return self.dq.pop()

# Queue (FIFO) using deque
class Queue:
    def __init__(self):
        self.dq = deque()
    
    def enqueue(self, x):
        self.dq.append(x)
    
    def dequeue(self):
        return self.dq.popleft()
```

### 2. **Sliding Window Problems**
```python
# Find maximum in every k-sized window
def maxSlidingWindow(nums, k):
    from collections import deque
    dq = deque()  # Stores indices
    result = []
    
    for i in range(len(nums)):
        # Remove elements outside window
        if dq and dq[0] < i - k + 1:
            dq.popleft()
        
        # Remove smaller elements
        while dq and nums[dq[-1]] < nums[i]:
            dq.pop()
        
        dq.append(i)
        if i >= k - 1:
            result.append(nums[dq[0]])
    
    return result

# Example: maxSlidingWindow([1,3,-1,-3,5,3,6,7], 3) = [3,3,5,5,6,7]
```

### 3. **BFS (Breadth-First Search) in Graphs**
```python
from collections import deque

def bfs(graph, start):
    queue = deque([start])
    visited = {start}
    
    while queue:
        node = queue.popleft()
        print(node, end=' ')
        
        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)
```

### 4. **Round-Robin Scheduling**
```python
def roundRobinScheduler(tasks, time_slice):
    task_queue = deque(tasks)
    
    while task_queue:
        task = task_queue.popleft()
        # Process task for time_slice
        if task.remaining_time > time_slice:
            task.remaining_time -= time_slice
            task_queue.append(task)  # Requeue if not done
```

### 5. **Palindrome Validation**
```python
def isPalindrome(s):
    dq = deque(s)
    
    while len(dq) > 1:
        if dq.popleft() != dq.pop():
            return False
    
    return True

# isPalindrome("racecar") = True
```

## ✅ Advantages of Deque

| Advantage | Explanation |
|-----------|-------------|
| **O(1) operations at both ends** | `append`, `appendleft`, `pop`, `popleft` are all O(1) |
| **Flexible** | Can act as both stack and queue |
| **Memory efficient** | No need for separate data structures for dual-end operations |
| **Thread-safe atomic operations** | Individual deque operations are atomic in CPython |
| **Built-in rotation** | `rotate()` is optimized for deque's internal structure |
| **Random access** | Can access elements by index in O(1) amortized |

---

## ❌ Disadvantages of Deque

| Disadvantage | Explanation |
|--------------|-------------|
| **Middle insertion/deletion is O(n)** | Use `insert()` at random positions sparingly |
| **Not optimal for sorting** | If sorting is needed, convert to list first |
| **Higher memory overhead** | Compared to a simple list for single-end access patterns |
| **Not thread-safe for compound operations** | Multiple operations: check then act patterns need external locking |
| **Limited search efficiency** | Searching requires O(n) linear scan |
| **No built-in slicing** | Unlike lists, deques don't support slice operations efficiently |

---

## 🎯 When to Use Deque

**Use Deque when:**
- ✅ You need O(1) operations at **both ends** (classic queue/stack scenarios)
- ✅ Implementing sliding window algorithms
- ✅ BFS/graph traversal
- ✅ Palindrome checking
- ✅ Undo/Redo functionality (with bounded history)
- ✅ Interleaving data from different sequences

**Don't use Deque when:**
- ❌ You only need access from one end (use simple list with `pop()`)
- ❌ You frequently need random access to middle elements
- ❌ You need thread-safe operations (use `queue.Queue`)
- ❌ You need to maintain sorted order frequently
- ❌ You need slicing operations (use list)

## 📊 Performance Comparison

```python
import time
from collections import deque

# List vs Deque for left-end operations
lst = list(range(100000))
dq = deque(range(100000))

# Left insertion (list is slow)
start = time.time()
for _ in range(1000):
    lst.insert(0, 1)
print(f"List insert(0): {time.time() - start:.4f}s")  # ~1.5s

# Left insertion (deque is fast)
start = time.time()
for _ in range(1000):
    dq.appendleft(1)
print(f"Deque appendleft: {time.time() - start:.4f}s")  # ~0.001s
```

**Result:** Deque is **1000x faster** for left-end operations!

---

## 🚫 Limitations

1. **No built-in thread safety:** Use `queue.Queue` or `queue.LifoQueue` for thread-safe operations
2. **No guaranteed O(1) for slicing:** Unlike lists
3. **No built-in sorting:** Must convert to list if sorting needed
4. **Fixed maxlen behavior:** When full, adding removes from opposite end silently
5. **Less CPU cache-friendly** than lists due to internal structure

---

## 📝 Conclusion

**Deque is a powerful, underutilized data structure** that excels in scenarios requiring access from both ends. It's the go-to choice for:
- Queue/Stack implementations
- Sliding window problems
- BFS algorithms
- Any scenario where you need O(1) operations at both ends

By understanding its strengths (O(1) dual-end operations, efficiency) and weaknesses (no sorting, middle operations are slow), you can make informed decisions about when to use deque in your DSA and system design solutions.

**Key Takeaway:** When in doubt, use deque for queue/stack problems and sliding window algorithms—it's specially optimized for these use cases in Python!

### 🔗 Quick Reference Cheat Sheet

```
Creation:           deque([1,2,3]) or deque()
Add rear:           dq.append(x)           → O(1)
Remove rear:        dq.pop()               → O(1)
Add front:          dq.appendleft(x)       → O(1)
Remove front:       dq.popleft()           → O(1)
Peek front:         dq[0]                  → O(1)
Peek rear:          dq[-1]                 → O(1)
Length:             len(dq)                → O(1)
Rotate right:       dq.rotate(n)           → O(n)
Reverse:            dq.reverse()           → O(n)
Extend rear:        dq.extend([...])       → O(k)
Extend front:       dq.extendleft([...])   → O(k)
Count element:      dq.count(x)            → O(n)
Clear:              dq.clear()             → O(n)
```

````python
from collections import deque

print("=" * 60)
print("1. BASIC OPERATIONS - Append & Pop from Rear")
print("=" * 60)

lst = deque([])

lst.append(100)
lst.append(200)
lst.append(300)
print(f"After append(100), append(200), append(300): {list(lst)}")

lst.appendleft(1)
lst.appendleft(9)
print(f"After appendleft(1), appendleft(9): {list(lst)}")

popped = lst.popleft()
print(f"After popleft(): Popped {popped}, Deque is {list(lst)}")

print("\n" + "=" * 60)
print("2. STACK IMPLEMENTATION")
print("=" * 60)

class Stack:
    def __init__(self):
        self.items = deque()
    
    def push(self, x):
        self.items.append(x)
    
    def pop(self):
        if not self.items:
            return None
        return self.items.pop()
    
    def peek(self):
        if not self.items:
            return None
        return self.items[-1]
    
    def is_empty(self):
        return len(self.items) == 0

stack = Stack()
for i in [10, 20, 30, 40]:
    stack.push(i)
print(f"Pushed 10, 20, 30, 40")
print(f"Peek: {stack.peek()}")
print(f"Pop: {stack.pop()}")
print(f"Pop: {stack.pop()}")

print("\n" + "=" * 60)
print("3. QUEUE IMPLEMENTATION")
print("=" * 60)

class Queue:
    def __init__(self):
        self.items = deque()
    
    def enqueue(self, x):
        self.items.append(x)
    
    def dequeue(self):
        if not self.items:
            return None
        return self.items.popleft()
    
    def is_empty(self):
        return len(self.items) == 0

queue = Queue()
for i in [100, 200, 300, 400]:
    queue.enqueue(i)
print(f"Enqueued 100, 200, 300, 400")
print(f"Dequeue: {queue.dequeue()}")
print(f"Dequeue: {queue.dequeue()}")
print(f"Is empty: {queue.is_empty()}")

print("\n" + "=" * 60)
print("4. ROTATION OPERATIONS")
print("=" * 60)

dq = deque([1, 2, 3, 4, 5])
print(f"Original: {list(dq)}")

dq.rotate(2)  # Rotate right by 2
print(f"After rotate(2) [right]: {list(dq)}")

dq.rotate(-2)  # Rotate left by 2
print(f"After rotate(-2) [left]: {list(dq)}")

print("\n" + "=" * 60)
print("5. EXTEND OPERATIONS")
print("=" * 60)

dq1 = deque([10, 20])
dq1.extend([30, 40, 50])
print(f"extend([30, 40, 50]) on [10, 20]: {list(dq1)}")

dq2 = deque([100])
dq2.extendleft([3, 2, 1])  # Note: order is reversed!
print(f"extendleft([3, 2, 1]) on [100]: {list(dq2)} (elements inserted in reverse)")

print("\n" + "=" * 60)
print("6. PALINDROME CHECK")
print("=" * 60)

def is_palindrome(s):
    dq = deque(s)
    while len(dq) > 1:
        if dq.popleft() != dq.pop():
            return False
    return True

test_strings = ["racecar", "hello", "madam", "python"]
for test in test_strings:
    result = is_palindrome(test)
    print(f"'{test}' is palindrome: {result}")

print("\n" + "=" * 60)
print("7. SLIDING WINDOW (Maximum in k-sized window)")
print("=" * 60)

def max_sliding_window(nums, k):
    dq = deque()
    result = []
    
    for i in range(len(nums)):
        # Remove elements outside current window
        if dq and dq[0] < i - k + 1:
            dq.popleft()
        
        # Remove smaller elements (maintaining decreasing order)
        while dq and nums[dq[-1]] < nums[i]:
            dq.pop()
        
        dq.append(i)
        
        # Add to result when window is full
        if i >= k - 1:
            result.append(nums[dq[0]])
    
    return result

nums = [1, 3, -1, -3, 5, 3, 6, 7]
k = 3
result = max_sliding_window(nums, k)
print(f"Array: {nums}")
print(f"Window size: {k}")
print(f"Max in each window: {result}")

print("\n" + "=" * 60)
print("8. PITFALL: maxlen BEHAVIOR")
print("=" * 60)

dq_limited = deque(maxlen=3)
print(f"Created deque with maxlen=3")
dq_limited.extend([1, 2, 3, 4, 5])
print(f"After extend([1,2,3,4,5]): {list(dq_limited)} (first elements discarded!)")

print("\n" + "=" * 60)
print("9. PERFORMANCE COMPARISON: List vs Deque")
print("=" * 60)

import time

# Test left-end insertion performance
lst = list(range(10000))
dq = deque(range(10000))

# List insert at index 0 (slow)
start = time.time()
for _ in range(100):
    lst.insert(0, 1)
list_time = time.time() - start

# Deque appendleft (fast)
start = time.time()
for _ in range(100):
    dq.appendleft(1)
deque_time = time.time() - start

print(f"Inserting at front 100 times:")
print(f"  List.insert(0, x):    {list_time:.6f}s")
print(f"  Deque.appendleft(x):  {deque_time:.6f}s")
print(f"  Speed ratio: {list_time / deque_time:.1f}x faster with deque!")
````

# Introduction to Functions
Functions are pre-defined blocks of code that perform a specific task. They help in code reusability.

**What are Functions?**

- **Built-in Functions:** Provided by Python (e.g., `print()`, `type()`).
- **User-Defined Functions:** Created by the user using the `def` keyword.

### Creating and Calling a Function
To define a function, use the `def` keyword followed by the function name and parentheses. Inside the function, write the block of code to execute. Call the function by using its name followed by parentheses.

````python
# Defining and calling a simple function
def greet():
    print("Hello, this is a greet function.")
    print("We are learning functions.")

# Call the function
greet()
````

### Functions vs Methods

- **Functions** are standalone blocks of code, like `print()` or custom functions.
- **Methods** are functions that belong to objects or data types, e.g., `list.append()` or `str.upper()`.

### Variable Scoping

Variables defined inside a function are **local** to that function and cannot be accessed outside. This prevents unintended interference with global variables.

````python
def addition():
    num1 = int(input("Enter num1: "))
    num2 = int(input("Enter num2: "))
    print(f"Answer is {num1 + num2}")

# num1 and num2 are local and not accessible outside the function
````

### Parameters and Arguments

- **Parameters:** Names listed in the function definition.
- **Arguments:** Actual values passed to the function when calling it.

````python
def add(n1, n2):  # n1 and n2 are parameters
    total = n1 + n2
    print(total)

add(5, 11)  # 5 and 11 are arguments
````

### Return Statement

Use `return` to send a value back to the caller. This allows the caller to store or use the result.

Printing only displays a value and does not give the caller access to it.

````python
def add(n1, n2):
    total = n1 + n2
    return total  # returns the value to the caller

# store returned value
result = add(10, 20)
print(result)  # Output: 30
````

# Advanced Function Concepts

### Positional and Named Arguments

By default, arguments are matched to parameters by position. You can also specify them by name, allowing any order and improving clarity.

````python
def total_marks(maths, science, english):
    total = maths + science + english
    print(total)

# call with named arguments in any order
total_marks(english=84, maths=89, science=74)
````

### Default Arguments

Assign default values to parameters so callers can omit them. The default is used when no argument is provided.

````python
def add(num1, num2=0):  # num2 has a default value of 0
    print(num1 + num2)

add(100)        # num2 defaults to 0; Output: 100
add(100, 200)   # num2 overridden; Output: 300
````

### `*args` and `**kwargs`

Python functions support flexible argument lists using `*args` and `**kwargs`.  

- `*args` collects **extra positional arguments** into a tuple. You can iterate over it or unpack it.
- `**kwargs` collects **extra keyword arguments** into a dictionary. Keys are argument names and values are the corresponding values.

These constructs are particularly useful when you don't know ahead of time how many arguments a caller may provide (for example, wrapper functions or decorators).

**Ordering rules** when defining parameters:
```python
def func(pos1, pos2, *args, kw1=1, **kwargs):
    ...
```
- Regular positional parameters first
- Then `*args` to capture any further positional arguments
- After that, default or keyword-only parameters
- Finally, `**kwargs` to capture extra named arguments

**Key points:**
- `args` will always be a tuple, even if no extra arguments are passed.
- `kwargs` will always be a dict, even if empty.
- You can unpack sequences or mappings into a function call using `*` and `**`, respectively.
- `*args` and `**kwargs` themselves are just conventional names; you could use other identifiers, but the asterisk syntax must be used.

````python
# examples covering *args and **kwargs

# simple *args usage
def add_all(*args):
    """Sum any number of positional arguments."""
    print("args type:", type(args))
    return sum(args)

print(add_all(1, 2, 3))           # 6
print(add_all(10, 20, 30, 40, 50)) # 150
print(add_all())                   # 0 (empty tuple)

# simple **kwargs usage
def describe_person(**kwargs):
    """Print keyword arguments in a friendly format."""
    print("kwargs type:", type(kwargs))
    for key, val in kwargs.items():
        print(f"{key} = {val}")


describe_person(name="Alice", age=30)

describe_person(city="Paris", profession="Artist", hobby="Painting")

# mixing regular parameters with *args and **kwargs
def mixed(a, b=2, *args, **kwargs):
    print("a=", a)
    print("b=", b)
    print("args=", args)
    print("kwargs=", kwargs)

mixed(10)
mixed(1, 3, 4, 5, x=100, y=200)

# unpacking sequences and mappings into function calls
numbers = [1, 2, 3]
print(add_all(*numbers))   # equivalent to add_all(1,2,3)

settings = {"debug": True, "timeout": 5}
def configure(**options):
    for k, v in options.items():
        print(f"{k} -> {v}")

configure(**settings)

# practical pattern: forwarding arguments in a wrapper

def logger(func, *args, **kwargs):
    print(f"Calling {func.__name__} with", args, kwargs)
    return func(*args, **kwargs)

# use correct decorator syntax
def logged(func):
    def wrapper(*args, **kwargs):
        print(f"Calling {func.__name__} with", args, kwargs)
        return func(*args, **kwargs)
    return wrapper

@logged

def multiply(x, y):
    return x * y

print(multiply(3, 4))
````

### Lambda Functions

Lambda functions are anonymous, one-line functions useful for simple operations.

Syntax: `lambda arguments: expression`

````python
# Normal function

def add_numbers(n1, n2, n3):
    return n1 + n2 + n3

# Equivalent lambda function
add_lambda = lambda n1, n2, n3: n1 + n2 + n3
print(add_lambda(10, 20, 30))  # Output: 60

# Take a argumnet which will be number and make a list 0 to N
make_list = lambda n: [i for i in range(0,n+1)]
print(make_list(9))

# Lambda for checking even/odd
check_even = lambda num: num % 2 == 0
print(check_even(6))  # Output: True
````

### Type Annotations

Type annotations provide hints about expected types for parameters and return values. They improve readability but are not enforced.

````python
def add(x: int, y: int) -> int:
    return x + y

from typing import List

def sum_of_list(x: List[int]) -> int:
    return sum(x)
````

### The `global` Keyword

- The `global` keyword lets a function assign to a variable defined at module scope (a global variable).
- Without `global`, assigning to a name inside a function creates a new local variable instead of modifying the module-level variable.
- Use `global` when you intentionally want to modify module-level state from inside a function. Prefer returning values and updating state at the call site to avoid unexpected side effects.

Common pitfalls:
- Attempting to modify a global variable without declaring it `global` inside the function leads to an `UnboundLocalError` when the variable is referenced before assignment.

Example summary below shows both the `global` behavior and a recommended alternative (returning new values).

````python
a = 15

def change():
    a = 30
    print(a)

def change_global():
    global a 
    a = 30
    print(a)

print(a)
change()
change_global()
print(a)
````

````python
# Demonstrations: `global` vs returning values

counter = 0

def increment_with_global():
    """Modify the module-level `counter` using `global`."""
    global counter
    counter += 1

def increment_without_global():
    """Attempting to modify without `global` raises UnboundLocalError when referencing before assignment."""
    try:
        counter += 1  # this will fail because Python treats `counter` as local
    except UnboundLocalError as e:
        print("UnboundLocalError:", e)

print("initial counter:", counter)
increment_with_global()
print("after increment_with_global:", counter)
increment_without_global()

# Recommended pattern: return the new value instead of mutating globals

def increment_return(c):
    return c + 1

counter = increment_return(counter)
print("after increment_return:", counter)
````

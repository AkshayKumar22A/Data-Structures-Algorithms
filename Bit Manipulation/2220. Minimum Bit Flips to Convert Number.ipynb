# Problem Analysis & Key Insights

## Problem Understanding

We need to find the **minimum number of bit flips** to convert one number to another.

**Key Insight:** A bit flip means changing a bit from 0→1 or 1→0. This happens exactly when the two numbers differ at that bit position.

## The Core Concept: XOR Operation

**XOR Property:** The XOR operation returns 1 when two bits are different and 0 when they're the same.

```
start = 10 = 1010
goal  =  7 = 0111
XOR:              
  1010
^ 0111
------
  1101  (result = 13)
```

- Bit 0: 0 ⊕ 1 = 1 (different, need flip) ✓
- Bit 1: 1 ⊕ 1 = 0 (same, no flip)
- Bit 2: 0 ⊕ 1 = 1 (different, need flip) ✓
- Bit 3: 1 ⊕ 0 = 1 (different, need flip) ✓

**Answer:** Count the 1s in the XOR result = 3 ✓

## Three Approaches to Count Set Bits

After getting the XOR result, we need to count the number of 1-bits. Here are three methods:

1. **Brian Kernighan's Algorithm** - Most efficient
2. **Bit Masking Loop (32 iterations)** - Fixed iterations
3. **Direct Bit Comparison** - Straightforward but verbose

````python
class Solution:
    def minBitFlips(self, start: int, goal: int) -> int:
        xor_flip = start^goal
        count = 0
        while xor_flip:
            count+=1
            xor_flip = xor_flip & (xor_flip-1)
        return count

print(Solution().minBitFlips(10,7))
````

# Solution 1: Brian Kernighan's Algorithm (Most Efficient ⭐)

## Intuition & Approach

**Key Idea:** Instead of checking every bit, we remove one set bit at a time using the formula `n & (n-1)`.

**Why `n & (n-1)` removes the rightmost set bit:**
- Subtracting 1 from a number flips all bits after the rightmost set bit
- ANDing clears the rightmost set bit

**Example:**
```
n =    1010 (10)
n-1 =  1001 (9)
n & (n-1) = 1000 (8) ← rightmost 1 removed!
```

**Algorithm:**
1. Calculate `xor_flip = start XOR goal` (positions where they differ)
2. While `xor_flip > 0`:
   - Increment counter
   - Remove rightmost set bit: `xor_flip = xor_flip & (xor_flip-1)`
3. Return count

## Code Explanation

```python
xor_flip = start^goal    # Get positions where bits differ
count = 0
while xor_flip:          # While there are set bits
    count += 1           # Count this set bit
    xor_flip = xor_flip & (xor_flip-1)  # Remove rightmost set bit
return count
```

**Each iteration removes exactly one set bit**, so loop runs k times (k = number of set bits).

## Detailed Dry Run: start=10, goal=7

**Step 1: Calculate XOR**
```
start = 10 = 1010
goal  =  7 = 0111
xor_flip = 1010 ^ 0111 = 1101 (13)
count = 0
```

**Iteration 1:**
```
xor_flip = 1101 (13)
Condition: 13 != 0 ✓ Enter loop
count = 1
xor_flip = xor_flip & (xor_flip - 1)
         = 1101 & 1100
         = 1100 (12)
Removed the rightmost 1 at position 0
```

**Iteration 2:**
```
xor_flip = 1100 (12)
Condition: 12 != 0 ✓ Continue
count = 2
xor_flip = 1100 & 1011
         = 1000 (8)
Removed the rightmost 1 at position 2
```

**Iteration 3:**
```
xor_flip = 1000 (8)
Condition: 8 != 0 ✓ Continue
count = 3
xor_flip = 1000 & 0111
         = 0000 (0)
Removed the rightmost 1 at position 3
```

**Iteration 4:**
```
xor_flip = 0 (0)
Condition: 0 == 0 ✗ Exit loop
Return count = 3 ✓
```

## Dry Run: start=3, goal=4

```
start = 3 = 0011
goal  = 4 = 0100
xor_flip = 0011 ^ 0100 = 0111 (7)

Iteration 1: 0111 & 0110 = 0110 (6) → count = 1
Iteration 2: 0110 & 0101 = 0100 (4) → count = 2
Iteration 3: 0100 & 0011 = 0000 (0) → count = 3
Exit loop, return 3 ✓
```

## Edge Cases

| Case | Input | XOR Result | Expected | Result |
|------|-------|-----------|----------|--------|
| **Same numbers** | start=5, goal=5 | 0000 | 0 | ✓ 0 |
| **Zero** | start=0, goal=0 | 0000 | 0 | ✓ 0 |
| **Start is 0** | start=0, goal=7 | 0111 | 3 | ✓ 3 |
| **Goal is 0** | start=7, goal=0 | 0111 | 3 | ✓ 3 |
| **All bits differ** | start=0, goal=$2^{30}-1$ | All 1s | 30 | ✓ |
| **Single bit** | start=1, goal=0 | 0001 | 1 | ✓ 1 |

**Why there are no negative cases:** Constraints guarantee 0 ≤ start, goal ≤ 10⁹ (only non-negative integers).

## Complexity Analysis

### Time Complexity

| Case | Complexity | Explanation |
|------|-----------|-------------|
| **Best** | $O(1)$ | start == goal → xor_flip = 0, loop doesn't run |
| **Average** | $O(\log n)$ | n bits, loop runs for number of set bits (≈ n/2) |
| **Worst** | $O(\log n)$ | All bits set → loop runs for all bits |

**Why $O(\log n)$?** The while loop runs k times where k is the number of set bits. For a 32-bit integer, k ≤ 32, so O(32) = **O(1)** (constant time). For general n-bit numbers, it's **O(log n)** or **O(k)** where k = popcount.

### Space Complexity

- **$O(1)$** - Only using `count` and `xor_flip` variables, no extra data structures

````python
class Solution:
    def minBitFlips(self, start: int, goal: int) -> int:
        ans = start^goal
        count = 0
        for i in range(0,32):
            if ans&(1<<i)!=0:
                count+=1
        return count

print(Solution().minBitFlips(10,7))
````

# Solution 2: Bit Masking with Fixed Iterations

## Intuition & Approach

**Key Idea:** Check each of the 32 bits individually using a mask `1 << i`.

**Algorithm:**
1. Calculate `ans = start XOR goal`
2. For each bit position i from 0 to 31:
   - Create mask: `1 << i` (a number with only bit i set)
   - Check if bit i is set: `(ans & (1 << i)) != 0`
   - If set, increment counter
3. Return count

**Why 32 iterations?** Since the constraint is 0 ≤ start, goal ≤ 10⁹, and $2^{30}$ = 1,073,741,824 > 10⁹, we need at most 30 bits. Using 32 covers all cases safely.

## Code Explanation

```python
ans = start ^ goal           # Get XOR result
count = 0
for i in range(0, 32):       # Check each of 32 bit positions
    if ans & (1 << i) != 0:  # If bit i is set
        count += 1           # Count it
return count
```

**For each i:**
- `1 << i` creates a mask with only bit i set (e.g., `1 << 2` = 0100)
- `ans & (1 << i)` extracts bit i (result is non-zero if bit i = 1)

## Detailed Dry Run: start=10, goal=7

**Step 1: Calculate XOR**
```
start = 10 = 1010
goal  =  7 = 0111
ans = 1010 ^ 0111 = 1101 (13)
count = 0
```

**Loop iterations:**
```
i=0:  mask = 1 << 0 = 0001 (1)
      ans & mask = 1101 & 0001 = 0001 ≠ 0 ✓ → count = 1

i=1:  mask = 1 << 1 = 0010 (2)
      ans & mask = 1101 & 0010 = 0000 = 0 ✗ → count = 1

i=2:  mask = 1 << 2 = 0100 (4)
      ans & mask = 1101 & 0100 = 0100 ≠ 0 ✓ → count = 2

i=3:  mask = 1 << 3 = 1000 (8)
      ans & mask = 1101 & 1000 = 1000 ≠ 0 ✓ → count = 3

i=4 to i=31: mask has bits beyond position 3
             ans & mask = 1101 & (high bits) = 0 ✗ → count stays 3
```

**Final Answer:** count = 3 ✓

## Dry Run: start=3, goal=4

```
start = 3 = 0011
goal  = 4 = 0100
ans = 0011 ^ 0100 = 0111 (7)

i=0: 0111 & 0001 = 0001 ✓ → count = 1
i=1: 0111 & 0010 = 0010 ✓ → count = 2
i=2: 0111 & 0100 = 0100 ✓ → count = 3
i=3 to 31: All 0 → count = 3
Return 3 ✓
```

## Edge Cases

| Case | Input | XOR | i where bits set | Count |
|------|-------|-----|-----------------|-------|
| **Same** | 5, 5 | 0 | None | ✓ 0 |
| **Zero** | 0, 0 | 0 | None | ✓ 0 |
| **Start=0** | 0, 7 | 7 | 0,1,2 | ✓ 3 |
| **Goal=0** | 7, 0 | 7 | 0,1,2 | ✓ 3 |
| **One bit diff** | 4, 5 | 1 | 0 | ✓ 1 |
| **All 32 bits** | 0, $2^{32}-1$ | All 1s up to bit 31 | 0-31 | ✓ 32 |

## Complexity Analysis

### Time Complexity

| Case | Complexity | Explanation |
|------|-----------|-------------|
| **Best** | $O(1)$ | Always 32 iterations regardless of input |
| **Average** | $O(1)$ | Always 32 iterations regardless of input |
| **Worst** | $O(1)$ | Always 32 iterations regardless of input |

**Why constant?** The loop always runs exactly 32 times, independent of the actual values or number of set bits. This is a **fixed-cost algorithm**.

### Space Complexity

- **$O(1)$** - Only using `ans` and `count` variables

````python
class Solution:
    def minBitFlips(self, start: int, goal: int) -> int:
        count = 0
        i = 0
        l1 = len(bin(start))
        l2 = len(bin(goal))
        while i <= l1 or i<= l2:
            mask = 1 << i
            if (start & mask) != (goal & mask):
                count+=1
            i+=1
        return count


print(Solution().minBitFlips(10,7))
````

# Solution 3: Direct Bit-by-Bit Comparison

## Intuition & Approach

**Key Idea:** Instead of XORing first, directly compare corresponding bits of both numbers using a mask.

**Algorithm:**
1. Initialize counter and loop variable
2. Get lengths of binary representations (for efficiency)
3. While i <= length of both numbers:
   - Create mask with only bit i set: `mask = 1 << i`
   - Extract bit i from both numbers
   - If bits differ: `(start & mask) != (goal & mask)`, increment counter
   - Move to next bit
4. Return count

**Advantage:** No intermediate XOR calculation needed, direct comparison.

## Code Explanation

```python
count = 0
i = 0
l1 = len(bin(start))  # Length of binary string representation
l2 = len(bin(goal))   # E.g., bin(10) = '0b1010', len = 6
while i <= l1 or i <= l2:
    mask = 1 << i
    if (start & mask) != (goal & mask):  # Bits at position i differ?
        count += 1
    i += 1
return count
```

**Important Detail:** `len(bin(n))` includes the '0b' prefix. For n=10, `len('0b1010')` = 6, representing 4 bits. This works because we're checking `i <= l1 or i <= l2`, ensuring we check all significant bits.

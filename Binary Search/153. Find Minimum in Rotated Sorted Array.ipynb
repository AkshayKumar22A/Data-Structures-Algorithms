# Solution Explanation: Find Minimum in Rotated Sorted Array (Binary Search)
 
## 1. Intuition & Approach
- The array is sorted and then rotated at an unknown pivot. All elements are unique.
- The minimum element is the pivot point where the order restarts.
- Instead of scanning every element, we use binary search to efficiently find the minimum.
- At each step, we check if the middle element is greater than the rightmost element:
    - If yes, the minimum is in the right half.
    - If no, the minimum is in the left half (including mid).
- We track the minimum seen so far using `ans = min(ans, nums[mid])`.
 
## 2. Code Explanation (step-by-step)
```python
from typing import List
class Solution:
    def findMin(self, nums: List[int]) -> int:
        ans = float("inf")
        low = 0
        high = len(nums) - 1
        while(low <= high):
            mid = low + (high - low) // 2
            ans = min(ans,nums[mid])
            if nums[mid] > nums[high]:
                low = mid + 1
            else:
                high = mid - 1
        return ans
```
- `ans` keeps track of the smallest value found so far.
- If `nums[mid] > nums[high]`, the minimum is in the right half, so move `low` to `mid + 1`.
- Otherwise, the minimum is in the left half (including `mid`), so move `high` to `mid - 1`.
- The loop continues until `low > high`, and `ans` holds the minimum value.
 
## 3. Dry Run (step-by-step for each iteration)
### Example: nums = [4, 5, 6, 7, 0, 1, 2]
| Step | low | high | mid | nums[mid] | ans | nums[mid] > nums[high]? | Next |
|------|-----|------|-----|-----------|-----|------------------------|------|
| 1    | 0   | 6    | 3   | 7         | 7   | Yes (7 > 2)            | low=4|
| 2    | 4   | 6    | 5   | 1         | 1   | No (1 <= 2)            | high=4|
| 3    | 4   | 4    | 4   | 0         | 0   | No (0 <= 0)            | high=3|
Loop ends. Final answer: 0 (minimum found at index 4).
 
### Example: nums = [1, 2, 3, 4, 5] (already sorted)
| Step | low | high | mid | nums[mid] | ans | nums[mid] > nums[high]? | Next |
|------|-----|------|-----|-----------|-----|------------------------|------|
| 1    | 0   | 4    | 2   | 3         | 3   | No (3 <= 5)            | high=1|
| 2    | 0   | 1    | 0   | 1         | 1   | No (1 <= 2)            | high=-1|
Loop ends. Final answer: 1 (minimum found at index 0).
 
## 4. Edge Cases
- **Single element:** [10] → returns 10.
- **Already sorted:** [1,2,3,4,5] → returns 1.
- **Negative numbers:** [-4,-3,-2,-1,0,1,2] rotated → works correctly.
- **Zero input:** [0,1,2,3] rotated → returns 0.
- **Large arrays:** Binary search ensures efficiency even for large n.
- **Empty array:** Not allowed by constraints (n >= 1).
 
## 5. Time & Space Complexity
- **Time Complexity:** O(log n) in all cases (best, average, worst) because each step halves the search space.
- **Space Complexity:** O(1) — only a few variables are used, no extra data structures.


````python
from typing import List
class Solution:
    def findMin(self, nums: List[int]) -> int:
        ans = float("inf")
        low = 0
        high = len(nums) - 1
        while(low <= high):
            mid = low + (high - low) // 2
            ans = min(ans,nums[mid])
            if nums[mid] > nums[high]:
                low = mid + 1
            else:
                high = mid - 1
        return ans    

result = Solution().findMin([4,5,6,7,0,1,2])
print(result)
````

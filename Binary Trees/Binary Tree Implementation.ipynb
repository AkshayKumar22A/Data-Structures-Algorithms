## Basic Node Relationships
- **Node:** The fundamental building block of a tree. It contains data (a value) and pointers to its left and right children.
- Root: The topmost node of the tree. It is the only node that does not have a parent. Every tree has exactly one root (unless it is empty).
- Parent: Any node that has at least one child node connected below it.
- Child: A node that is connected directly beneath another node (its parent). In a binary tree, a parent can have at most a Left Child and a Right Child.
- Siblings: Nodes that share the exact same parent.
- Leaf Node (or External Node): A node at the very bottom of the tree that has zero children (both left and right pointers are None).
- Internal Node: Any node that has at least one child. (Every node except the leaf nodes).

## Measurements and Metrics
- Edge: The connecting link between two nodes. If a tree has $N$ nodes, it will always have exactly $N - 1$ edges.
- Path: A continuous sequence of edges and nodes connecting one node to another.
- Depth (of a Node): The number of edges along the path from the root node down to that specific node. The root node has a depth of 0.
- Height (of a Node): The number of edges on the longest path from that specific node down to a leaf. Leaf nodes have a height of 0.
- Height (of a Tree): The height of its root node. (It represents the maximum depth of any node in the tree).
- Level: All nodes that are situated at the same depth. The root is at Level 0, its children are at Level 1, their children at Level 2, and so on.

## Structural Concepts
- Subtree: A tree entirely contained within a larger tree. Every node in a binary tree acts as the root of its own left subtree and right subtree.
- Ancestor: Any node situated on the path from the root down to a given node. (e.g., A node's parent, grandparent, etc., are its ancestors).
- Descendant: Any node situated along a path originating from a given node down to the leaves. (e.g., A node's children, grandchildren, etc., are its descendants).

## Types of Binary Trees
1. Full Binary Tree (Strict or Proper Binary Tree): In a Full Binary Tree, every single node has exactly 0 or 2 children. You will never see a node with just one child.
    - Rule of thumb: Nodes either have a full set of children, or they are leaves.

2. Complete Binary Tree: A Complete Binary Tree is about how the tree is "filled in." All levels of the tree must be completely filled with nodes, except possibly for the very last level. Furthermore, the last level must be filled strictly from left to right, with no gaps. 
    - Why it matters: This structure is the foundation of the Heap data structure (like a Min-Heap or Max-Heap). Because there are no gaps, complete binary trees can be efficiently represented using standard arrays rather than node objects.

3. Perfect Binary Tree: A Perfect Binary Tree is both full and complete. All internal nodes have exactly two children, and every single leaf node is at the exact same level (or depth).
    - Key property: If a perfect binary tree has height $h$, it has exactly $2^{h+1} - 1$ total nodes. It is perfectly symmetrical.

4. Balanced Binary Tree: A tree is considered balanced if the height of the left subtree and the right subtree of any node differ by no more than 1.
    - Why it matters: In unbalanced trees (like a tree where every node only has a right child), traversing the tree becomes slow ($O(N)$ time). Balanced trees (like AVL trees or Red-Black trees) guarantee that search, insertion, and deletion operations stay fast at $O(\log N)$ time.
5. Binary Search Tree (BST): A Binary Search Tree dictates how the values inside the nodes are arranged, rather than just the structural shape. It follows a strict sorting rule:
    - The value of the left child is always strictly less than the parent's value.
    - The value of the right child is always strictly greater than the parent's value.
    - This rule applies recursively to all subtrees.

# Different Traversals
1. DFS [Depth First Search] (Recursion)
    - InOrder -> Left, Root, Right.
        - Best Use Case: If you perform an InOrder traversal on a Binary Search Tree (BST), it will visit the nodes in strictly sorted, ascending order.
    - PreOrder -> Root, Left, Right.
        - Best Use Case: PreOrder is perfect for copying or cloning a tree. Because you process the parent first, you can easily create the root of your new tree and then attach the cloned left and right children to it.
    - PostOrder -> Left, Right, Root.
        - Best Use Case: PostOrder is used for deleting a tree or freeing up memory. You must delete the children before you delete the parent, making a bottom-up approach necessary.

2. BFS [Breadth Frst Search] (Loop and Queue(FIFO))
    - Level Ordr Traversals
        - Best Use Case: BFS is ideal for finding the shortest path to a specific node, or for questions that ask you to look at the tree from a specific angle (like finding the "Right Side View" of a binary tree).

```python
from collections import deque

# 1. The Node Class
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

# 2. The Traversals Class
class TreeTraversals:
    
    # --- DFS Traversals (Recursive) ---
    # 1. Preorder: Root -> Left -> Right
    # Use case: Creating a copy of the tree, prefix notation.
    def preorder(self, root: TreeNode) -> list[int]:
        """Root -> Left -> Right"""
        if not root:
            return []
        return [root.val] + self.preorder(root.left) + self.preorder(root.right)

    # 2. Inorder: Left -> Root -> Right
    # Use case: Flattens a Binary Search Tree (BST) into sorted order.
    def inorder(self, root: TreeNode) -> list[int]:
        """Left -> Root -> Right"""
        if not root:
            return []
        return self.inorder(root.left) + [root.val] + self.inorder(root.right)

    # 3. Postorder: Left -> Right -> Root
    # Use case: Deleting a tree, processing children before the parent.
    def postorder(self, root: TreeNode) -> list[int]:
        """Left -> Right -> Root"""
        if not root:
            return []
        return self.postorder(root.left) + self.postorder(root.right) + [root.val]

    # --- BFS Traversal (Iterative) ---
    # Level Order Traversal
    # Use case: Finding the shortest path, right-side view of a tree.
    def levelOrder(self, root: TreeNode) -> list[list[int]]:
        """Level by Level (Top to Bottom, Left to Right)"""
        if not root:
            return []
        
        result = []
        queue = deque([root])
        
        while queue:
            level_size = len(queue)
            current_level = []
            
            # Process all nodes at the current depth level
            for _ in range(level_size):
                node = queue.popleft()
                current_level.append(node.val)
                
                if node.left:
                    queue.append(node.left)
                if node.right:
                    queue.append(node.right)
            
            # Append the current level's values as a sub-list
            result.append(current_level)
            
        return result
    
    def levelOrder1(self, root: TreeNode) -> list[list[int]]:
        """Level by Level (Top to Bottom, Left to Right)"""
        if not root:
            return []
        
        result = []
        queue = deque([root])
        
        while queue:
            node = queue.popleft()
            result.append(node.val)
            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)
        return result

# 3. Execution and Testing Block
if __name__ == "__main__":
    # Build the dummy tree:
    #         1
    #        / \
    #       2   3
    #      / \   \
    #     4   5   6
    
    node1 = TreeNode(1)
    node2 = TreeNode(2)
    node3 = TreeNode(3)
    node4 = TreeNode(4)
    node5 = TreeNode(5)
    node6 = TreeNode(6)

    # Link nodes
    node1.left = node2
    node1.right = node3
    node2.left = node4
    node2.right = node5
    node3.right = node6

    root = node1 # The entry point of our tree

    # Instantiate the traversals class
    traversals = TreeTraversals()

    # Run and print the results
    print("--- Depth-First Search (DFS) ---")
    print(f"PreOrder  (Root, Left, Right): {traversals.preorder(root)}")
    print(f"InOrder   (Left, Root, Right): {traversals.inorder(root)}")
    print(f"PostOrder (Left, Right, Root): {traversals.postorder(root)}")
    
    print("\n--- Breadth-First Search (BFS) ---")
    print("Level Order (Level by Level):")
    bfs_result = traversals.levelOrder(root)
    for i, level in enumerate(bfs_result):
        print(f"Level Order   Level {i}: {level}")
    
    print(f"Level Order traversals: {traversals.levelOrder1(root)}")
```

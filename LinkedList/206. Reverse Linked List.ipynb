````python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def reverseList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        temp = head
        stack = []
        while temp:
            stack.append(temp.val)
            temp = temp.next

        temp = head
        while temp:
            temp.val = stack.pop()
            temp = temp.next
        
        return head
````

## **APPROACH 1: Stack-Based Approach**

### **Intuition and Approach**
The idea is simple:
1. **Collect all values**: Traverse the linked list and push all node values onto a stack
2. **Reassign values**: Traverse the list again and pop values from the stack, assigning them back to nodes in reverse order
3. Since a stack follows LIFO (Last-In-First-Out), popping values gives us elements in reverse order

This is a **non-pointer manipulation** approach that modifies node values rather than restructuring the list.

### **Code Explanation in Detail**

- **Line 1**: `temp = head` - Create a temporary pointer pointing to the first node
- **Lines 3-4**: First while loop traverses from head to end, pushing each node's value onto the stack
- **Lines 6-7**: Second while loop traverses the list again, popping values from stack (reverse order) back to nodes
- **Line 8**: Return the same head pointer (the list structure remains unchanged, only values are reversed)

### **Dry Run in Detail**

Given: `head = [1, 2, 3, 4, 5]`

**PHASE 1 - Building Stack:**

| Iteration | temp.val | Stack After Push | temp moves to |
|-----------|----------|------------------|---------------|
| 1 | 1 | [1] | Node 2 |
| 2 | 2 | [1, 2] | Node 3 |
| 3 | 3 | [1, 2, 3] | Node 4 |
| 4 | 4 | [1, 2, 3, 4] | Node 5 |
| 5 | 5 | [1, 2, 3, 4, 5] | None |
| Exit | - | [1, 2, 3, 4, 5] | - |

**PHASE 2 - Popping and Reassigning:**

| Iteration | temp.val (before) | stack.pop() | temp.val (after) | temp moves to |
|-----------|---|---|---|---|
| 1 | 1 | 5 | 5 | Node 2 |
| 2 | 2 | 4 | 4 | Node 3 |
| 3 | 3 | 3 | 3 | Node 4 |
| 4 | 4 | 2 | 2 | Node 5 |
| 5 | 5 | 1 | 1 | None |
| Exit | - | - | - | - |

**Final List**: `[5, 4, 3, 2, 1]` ✓

### **Edge Cases**

1. **Empty list** (`head = []`):
   - First while loop: doesn't execute (temp is None)
   - Stack remains `[]`
   - Second while loop: doesn't execute
   - Returns `None` ✓

2. **Single node** (`head = [1]`):
   - Stack: `[1]`
   - After popping: node value = 1 (unchanged)
   - Returns single node ✓

3. **Two nodes** (`head = [1, 2]`):
   - Stack: `[1, 2]`
   - Pop order: 2, then 1
   - Result: `[2, 1]` ✓

### **Time and Space Complexity**

| Metric | Best Case | Average Case | Worst Case |
|--------|-----------|--------------|-----------|
| **Time** | O(n) | O(n) | O(n) |
| **Space** | O(n) | O(n) | O(n) |

- **Time**: O(n) - Two separate traversals of the list, each takes O(n)
- **Space**: O(n) - Stack stores all n node values

````python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def reverseList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        curr = head
        Nextcurr = None
        prev = None
        
        while curr:
            # Change connection
            Nextcurr = curr.next
            curr.next = prev
            # Move Pointers
            prev = curr
            curr = Nextcurr

        return prev
````

## **APPROACH 2: Iterative Pointer Reversal (OPTIMAL)**

### **Intuition and Approach**

This approach **reverses pointers** without using extra space:
1. Use three pointers: `prev`, `curr`, and `Nextcurr`
2. At each step, **reverse the direction** of the current node's pointer
3. Move all pointers one step forward
4. Continue until we reach the end

This is the **optimal pointer-based approach** - it reverses the actual list structure.

### **Code Explanation in Detail**

- **Line 1**: `curr = head` - Start at the head node
- **Line 2**: `Nextcurr = None` - Placeholder for the next node to process
- **Line 3**: `prev = None` - Start with no previous node (will be tail initially)
- **Line 4**: `Nextcurr = curr.next` - **Critical step**: Save the next node because we're about to change curr.next
- **Line 5**: `curr.next = prev` - **Reverse the pointer**: Current node now points to previous
- **Line 6**: `prev = curr` - Move prev forward
- **Line 7**: `curr = Nextcurr` - Move curr forward using saved next node
- **Line 8**: Return `prev` (not head!) because prev is now at the tail which becomes the new head

### **Dry Run in Detail**

Given: `head = [1, 2, 3]`

**Initial State:**
```
prev: None
curr: Node(1) → Node(2) → Node(3) → None
Nextcurr: None
```

**Iteration 1:**
- `Nextcurr = curr.next` → Node(2)
- `curr.next = prev` → Node(1).next = None
- `prev = curr` → prev points to Node(1)
- `curr = Nextcurr` → curr points to Node(2)

```
List: Node(1) ← ... (broken temporarily)
      Node(2) → Node(3) → None
prev: Node(1)
curr: Node(2)
```

**Iteration 2:**
- `Nextcurr = curr.next` → Node(3)
- `curr.next = prev` → Node(2).next = Node(1)
- `prev = curr` → prev points to Node(2)
- `curr = Nextcurr` → curr points to Node(3)

```
List: None ← Node(1) ← Node(2)
             Node(3) → None
prev: Node(2)
curr: Node(3)
```

**Iteration 3:**
- `Nextcurr = curr.next` → None
- `curr.next = prev` → Node(3).next = Node(2)
- `prev = curr` → prev points to Node(3)
- `curr = Nextcurr` → curr = None (exit loop)

```
List: None ← Node(1) ← Node(2) ← Node(3)
prev: Node(3) ← NEW HEAD
curr: None
```

**Final Result**: Return `prev` (Node(3))
- Traversing from Node(3): 3 → 2 → 1 → None ✓

### **Edge Cases**

1. **Empty list** (`head = None`):
   - `curr = None` from the start
   - While loop doesn't execute
   - Returns `prev = None` ✓

2. **Single node** (`head = [5]`):
   - Iteration 1:
     - `Nextcurr = None`
     - `curr.next = None` (unchanged)
     - `prev = Node(5)`
     - `curr = None` (exit)
   - Returns Node(5) ✓

3. **Negative values** (`head = [-5, -10, -15]`):
   - The algorithm works identically - values don't affect pointer logic
   - Result: `[-15, -10, -5]` ✓

4. **Large list** (5000 nodes - max constraint):
   - O(n) time, O(1) space - handles efficiently ✓

### **Time and Space Complexity**

| Metric | Best Case | Average Case | Worst Case |
|--------|-----------|--------------|-----------|
| **Time** | O(n) | O(n) | O(n) |
| **Space** | O(1) | O(1) | O(1) |

- **Time**: O(n) - Single traversal of all n nodes
- **Space**: O(1) - Only three pointers used, constant space regardless of list size

````python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def reverseList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        # Base case: empty list or single node
        if not head or not head.next:
            return head
        
        # Recursive case: reverse the rest of the list
        new_head = self.reverseList(head.next)
        
        # Reverse the link between current and next node
        head.next.next = head
        head.next = None
        
        return new_head
````

## **APPROACH 3: Recursive Approach**

### **Intuition and Approach**

The recursive approach works on a simple principle:
1. **Base case**: If the list is empty or has only one node, it's already reversed
2. **Recursive case**: 
   - Recursively reverse the rest of the list (from the next node onwards)
   - Once the recursion returns, we know the rest of the list is reversed
   - Now reverse the link between the current node and the next node
   - Make the current node the tail (point to None)

The key insight: The recursion reaches the last node first, which becomes the new head. As the recursion unwinds, each node reverses its pointer to the previous node.

### **Code Explanation in Detail**

```python
class Solution:
    def reverseList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        # Base case: empty list or single node
        if not head or not head.next:
            return head
        
        # Recursive case: reverse the rest of the list
        new_head = self.reverseList(head.next)
        
        # Reverse the link between current and next node
        head.next.next = head
        head.next = None
        
        return new_head
```

**Line-by-line breakdown:**

- **Lines 1-2**: Base case - if list is empty (`not head`) or single node (`not head.next`), return it as is. This is the anchor point where recursion stops.

- **Line 4**: `new_head = self.reverseList(head.next)` - Recursively reverse the **rest** of the list starting from the next node. This goes all the way to the last node.

- **Line 6**: `head.next.next = head` - **Reverse the pointer**: The next node's next pointer now points back to the current node. This "flips" the connection.
  - Before: `current → next → ...`
  - After: `current ← next` (in terms of the relationship)

- **Line 7**: `head.next = None` - Make the current node's next pointer point to None. This ensures the node becomes a tail in the reversed list.

- **Line 8**: Return `new_head` - The new head (original tail) is returned at every level.

### **Dry Run in Detail**

Given: `head = [1, 2, 3]`

**Call Stack (Going Down - Forward Recursion):**

```
Call 1: reverseList(Node(1))
  ↓ calls
Call 2: reverseList(Node(2))
  ↓ calls
Call 3: reverseList(Node(3))
  ↓ calls
Call 4: reverseList(None) → Base case returns Node(3)
```

**Call Stack (Coming Back - Unwinding Recursion):**

**Call 3 (Processing Node(3)):**
- `head = Node(3)`, `head.next = None`
- Base case: `not head.next` is True → return Node(3)
- `new_head = Node(3)`
- Returns Node(3) to Call 2

**Call 2 (Processing Node(2)):**
- `head = Node(2)`, `head.next = Node(3)`
- `new_head = self.reverseList(Node(3))` → returns Node(3)
- `head.next.next = head` → `Node(3).next = Node(2)`
  - Now: `Node(2) ← Node(3)`
- `head.next = None` → `Node(2).next = None`
  - Now: `Node(2)` points to None
- List state: `None ← Node(2) ← Node(3)`
- Returns Node(3) to Call 1

**Call 1 (Processing Node(1)):**
- `head = Node(1)`, `head.next = Node(2)`
- `new_head = self.reverseList(Node(2))` → returns Node(3)
- `head.next.next = head` → `Node(2).next = Node(1)`
  - Now: `Node(1) ← Node(2)`
- `head.next = None` → `Node(1).next = None`
  - Now: `Node(1)` points to None
- List state: `None ← Node(1) ← Node(2) ← Node(3)`
- Returns Node(3) as final answer

**Final Result**: `3 → 2 → 1 → None` ✓

**Visual representation of list transformation:**

```
Initial:     1 → 2 → 3 → None

After Call 3:  1 → 2 → 3 (no change yet, base case)

After Call 2:  1 → 2 ← 3
                    ↓
                   None

After Call 1:  1 ← 2 ← 3
               ↓
              None

Final:       3 → 2 → 1 → None
```

### **Edge Cases**

1. **Empty list** (`head = None`):
   - Base case: `not head` is True
   - Returns None immediately ✓

2. **Single node** (`head = [5]`):
   - Base case: `not head.next` is True
   - Returns Node(5) directly ✓

3. **Two nodes** (`head = [1, 2]`):
   - Call 1: reverseList(Node(1))
   - Call 2: reverseList(Node(2)) → Base case, returns Node(2)
   - Back in Call 1:
     - `Node(2).next = Node(1)`
     - `Node(1).next = None`
   - Returns Node(2)
   - Result: `2 → 1 → None` ✓

4. **Negative values** (`head = [-10, -5, 0, 5]`):
   - The algorithm works identically - values don't affect recursion
   - Result: `[5, 0, -5, -10]` ✓

5. **Circular reference prevention**:
   - The `head.next = None` line prevents any cycles from forming ✓

### **Time and Space Complexity**

| Metric | Best Case | Average Case | Worst Case |
|--------|-----------|--------------|-----------|
| **Time** | O(n) | O(n) | O(n) |
| **Space** | O(n) | O(n) | O(n) |

**Time Complexity**: O(n) - Each node is visited once during recursion

**Space Complexity**: O(n) - **Call stack space**. The recursion goes n levels deep:
- For a list of 5 nodes, the call stack has 5 frames at maximum depth
- Unlike iterative approaches, recursive calls occupy stack memory
- This is the main drawback of the recursive approach

### **Comparison: All Three Approaches**

| Feature | Approach 1 (Stack) | Approach 2 (Iterative) | Approach 3 (Recursive) |
|---------|-------------------|---------------------|-----------------------|
| **Time** | O(n) | O(n) | O(n) |
| **Space** | O(n) - heap | O(1) | O(n) - stack |
| **Modifies Values** | Yes | No | No |
| **Modifies Pointers** | No | Yes | Yes |
| **Code Complexity** | Easy | Moderate | Moderate |
| **Risk of Stack Overflow** | No | No | Yes (large lists) |
| **Recommended** | Learning | Production | Interview showcase |

### **Key Takeaway**

- **Approach 2 (Iterative)** is the most efficient for production code (O(1) space, no stack overflow risk)
- **Approach 3 (Recursive)** is elegant and demonstrates understanding of recursion, but uses O(n) stack space
- **Approach 1 (Stack)** is intuitive for learning but uses heap space unnecessarily

````python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def deleteMiddle(self, head: Optional[ListNode]) -> Optional[ListNode]:
        temp = head
        count = -1
        while temp:
            count+=1
            temp = temp.next
        count = count / 2
        count = math.ceil(count)

        if count == 0:
            return None
        
        temp = head
        i = 0
        while temp:
            if i == count-1:
                break
            i+=1
            temp = temp.next

        temp.next = temp.next.next
        return head
        
````

## **SOLUTION 1: Count-Based Approach (Two-Pass)**

### **1.1 Intuition and Approach**

**Core Idea:**
- First, find the total length of the list by traversing it completely.
- Calculate the middle index: `middle = ⌊length / 2⌋`
- If the middle index is 0 (only 1 node), return `None` (delete the head).
- Otherwise, traverse to the node **before the middle** and skip the middle node by updating pointers.

**Why it works:**
- Knowing the total length makes it straightforward to find the middle position.
- Two-pass approach: collect info (pass 1) then act on that info (pass 2).
- Works for all list sizes and configurations.

### **1.2 Code Explanation in Detail**

```python
class Solution:
    def deleteMiddle(self, head: Optional[ListNode]) -> Optional[ListNode]:
        temp = head
        count = -1                          # Start at -1, will become length after loop
        
        # Pass 1: Count total nodes
        while temp:
            count += 1
            temp = temp.next
        
        # Calculate middle index using ceil(count/2)
        count = count / 2
        count = math.ceil(count)            # Ensure integer middle index
        
        # Special case: single node list
        if count == 0:
            return None
        
        # Pass 2: Find node before middle
        temp = head
        i = 0
        while temp:
            if i == count - 1:              # Found node before middle
                break
            i += 1
            temp = temp.next
        
        # Remove middle node by skipping it
        temp.next = temp.next.next
        return head
```

````python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def deleteMiddle(self, head: Optional[ListNode]) -> Optional[ListNode]:
        # Edge Case: If list has only 1 node, delete it and return None
        if not head.next:
            return None
            
        slow = head
        fast = head.next.next # Start fast 2 steps ahead!
        
        # Why start fast at head.next.next?
        # We want 'slow' to stop ONE STEP BEFORE the middle.
        
        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next
            
        # Now 'slow' is exactly behind the middle node.
        # Skip the middle node:
        slow.next = slow.next.next
        
        return head
````

## **SOLUTION 2: Two-Pointer Approach (One-Pass)**

### **2.1 Intuition and Approach**

**Core Idea:**
- Use two pointers (`slow` and `fast`) with a gap of exactly 2 nodes.
- Start `fast` at `head.next.next` (two nodes ahead of head).
- Advance both pointers in sync: slow by 1, fast by 2.
- When `fast` reaches the end, `slow` will be at the node **before the middle**.
- Skip the middle node by updating pointers.

**Why start fast at head.next.next:**
- A pointer starting 2 steps ahead, when advanced by 2 each iteration, maintains a gap.
- This gap ensures that when fast reaches the end (or becomes None), slow is positioned exactly before the middle.
- Handles both odd and even-length lists correctly.

**Key insight:**
- For any length n, advancing both by maintaining a 2-node gap will place slow one step before the middle.
- This is more efficient than counting first.

### **2.2 Code Explanation in Detail**

```python
class Solution:
    def deleteMiddle(self, head: Optional[ListNode]) -> Optional[ListNode]:
        # Edge case: single node list
        if not head.next:
            return None
        
        # Initialize pointers
        slow = head                         # Start at head
        fast = head.next.next               # Start 2 steps ahead
        
        # Move both pointers until fast reaches end
        while fast and fast.next:
            slow = slow.next                # Advance slow by 1
            fast = fast.next.next           # Advance fast by 2
        
        # Now slow is before the middle node
        # Remove the middle node
        slow.next = slow.next.next
        
        return head
```

**Line-by-Line Breakdown:**

| Line | Code | Explanation |
|------|------|-------------|
| 2-3 | `if not head.next:` | Check if only 1 node exists |
| 4 | `return None` | Single node → delete it → return None |
| 7 | `slow = head` | Start slow at head |
| 8 | `fast = head.next.next` | Start fast 2 steps ahead |
| 11 | `while fast and fast.next:` | Continue while fast can move 2 more steps |
| 12 | `slow = slow.next` | Move slow by 1 step |
| 13 | `fast = fast.next.next` | Move fast by 2 steps |
| 17 | `slow.next = slow.next.next` | Skip middle node |
| 19 | `return head` | Return modified list |

````python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def removeNthFromEnd(self, head: Optional[ListNode], n: int) -> Optional[ListNode]:
        length = 0
        temp = head
        while temp:
            length+=1
            temp = temp.next
        
        if length == n:
            new_head = head.next
            return new_head
        
        remove = length - n
        temp = head

        while temp is not None:
            remove -= 1
            if remove == 0:
                break
            temp = temp.next

        # Remove the nth node from the end
        temp.next = temp.next.next

        return head
````

## **SOLUTION 1: Two-Pass Approach (Length-Based)**

### **1.1 Intuition and Approach**

**Problem Statement:**
Given a linked list, remove the nth node from the end of the list and return the head. For example, in list [1,2,3,4,5] with n=2, remove node 4 to get [1,2,3,5].

**Key Insight:**
If we know the total length of the list, we can calculate the position from the beginning: `position = length - n`. Then we can traverse to that position and remove the node.

**Approach:**
1. **Pass 1:** Count total number of nodes in the list
2. Handle special case: if n equals the list length, the head node must be removed
3. **Pass 2:** Calculate the node before the target (`length - n` position)
4. Traverse to that node and update pointers to skip the target node
5. Return the head

**Why This Works:**
- Converting "nth from end" to "position from start" is straightforward: `from_start = length - n`
- Two passes guarantee we have all information needed
- Works for all cases including removing the head node

### **1.2 Code Explanation in Detail**

````python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def removeNthFromEnd(self, head: Optional[ListNode], n: int) -> Optional[ListNode]:
        slow,fast = head,head
        for _ in range(n):
            fast = fast.next
        
        if fast == None:
            return head.next
        
        while fast.next:
            fast = fast.next
            slow = slow.next
        slow.next = slow.next.next
        return head
````


## **SOLUTION 2: Two-Pointer Approach (One-Pass) ⭐ OPTIMAL**

### **2.1 Intuition and Approach**

**Key Insight:**
Use two pointers with a gap of n nodes between them. When the fast pointer reaches the end, the slow pointer will be exactly at the node before the target.

**Approach:**
1. Initialize two pointers (`slow` and `fast`) at the head
2. Move `fast` pointer n steps ahead
3. If `fast` becomes None after n steps, the head node is the target (remove it)
4. Move both pointers simultaneously until `fast.next` is None
5. At this point, `slow` is at the node before target; remove it
6. Return head

**Why This Works:**
- Creates a "sliding window" of n nodes between the pointers
- When fast reaches the end, slow is perfectly positioned
- Single pass through the list
- Elegant and efficient

### **2.2 Code Explanation in Detail**

```python
class Solution:
    def removeNthFromEnd(self, head: Optional[ListNode], n: int) -> Optional[ListNode]:
        # Initialize both pointers at head
        slow, fast = head, head
        
        # Move fast pointer n steps ahead
        for _ in range(n):
            fast = fast.next
        
        # If fast is None, we're removing the head
        if fast == None:
            return head.next
        
        # Move both pointers until fast reaches the last node
        while fast.next:
            fast = fast.next
            slow = slow.next
        
        # Remove the target node by skipping it
        slow.next = slow.next.next
        
        return head
```

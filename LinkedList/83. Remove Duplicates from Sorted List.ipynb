````python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next

class Solution:
    def deleteDuplicates(self, head: Optional[ListNode]) -> Optional[ListNode]:
        if head is None:
            return head

        temp = head
        while temp and temp.next:
            if temp.val == temp.next.val:
                temp.next = temp.next.next
            else:
                temp = temp.next
        
        return head
````

## **1. INTUITION AND APPROACH**

**Problem Statement:**
Given a sorted linked list, remove all duplicate consecutive nodes so that each unique value appears only once, while maintaining the sorted order.

**Key Insight:**
Since the linked list is **already sorted**, duplicate elements will always be consecutive. This means we can efficiently identify and remove duplicates by comparing adjacent nodes.

**Approach:**
1. Start at the head of the linked list
2. Compare each node's value with the next node's value
3. If they are equal (duplicate found), skip the duplicate by pointing to the node after it
4. If they are different, move to the next node
5. Continue until we reach the end of the list

**Why This Works:**
- Linear single-pass traversal is possible because duplicates are guaranteed to be adjacent
- By directly manipulating the `next` pointer, we avoid creating new nodes and effectively "delete" duplicates

## **2. CODE EXPLANATION IN DETAIL**

```python
class Solution:
    def deleteDuplicates(self, head: Optional[ListNode]) -> Optional[ListNode]:
        # Step 1: Handle empty list
        if head is None:
            return head
        
        # Step 2: Initialize temp pointer at head
        temp = head
        
        # Step 3: Traverse the list
        while temp and temp.next:
            # Check if current node's value equals next node's value
            if temp.val == temp.next.val:
                # If duplicate found, skip the duplicate node
                # by connecting current node to the node after duplicate
                temp.next = temp.next.next
            else:
                # If not duplicate, move to next node
                temp = temp.next
        
        # Step 4: Return the modified head
        return head
```

**Line-by-Line Breakdown:**

| Line | Code | Explanation |
|------|------|-------------|
| 1-2 | `if head is None: return head` | Edge case: if list is empty, return None |
| 4 | `temp = head` | Create a pointer starting at the head |
| 6 | `while temp and temp.next:` | Continue while current node exists AND next node exists |
| 8 | `if temp.val == temp.next.val:` | Check if duplicate found |
| 10 | `temp.next = temp.next.next` | **Skip** the duplicate by jumping over it |
| 12 | `temp = temp.next` | Move to next node (only if not duplicate) |
| 15 | `return head` | Return the modified list |

## **5. TIME AND SPACE COMPLEXITY**

### **Time Complexity:**

| Case | Complexity | Explanation |
|------|-----------|-------------|
| **Best Case** | O(n) | List with no duplicates; must check every node once |
| **Average Case** | O(n) | Must traverse entire list regardless of duplicates |
| **Worst Case** | O(n) | All elements are duplicates; still need to check each node |

**Overall: O(n)** where n is the number of nodes in the linked list.

**Analysis:**
- The while loop runs through each node exactly once
- Each iteration performs constant time operations (comparison, pointer update)
- We visit each node at most once: O(1) × n = O(n)

### **Space Complexity:**

| Case | Complexity | Explanation |
|------|-----------|-------------|
| **Best Case** | O(1) | No additional data structures |
| **Average Case** | O(1) | Only using `temp` pointer variable |
| **Worst Case** | O(1) | Constant auxiliary space regardless of input |

**Overall: O(1) - Constant Space**

**Analysis:**
- No additional arrays, hash maps, or recursive call stack
- Only one pointer variable `temp` is used
- The modification happens in-place on the existing list
- Memory usage does not scale with input size

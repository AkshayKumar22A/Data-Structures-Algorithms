Methods to Implement:
prepend(data): Add a node to the start.
append(data): Add a node to the end.
insert_at_index(index, data): Insert a node at the specified 0-based index.
delete_head(): Remove the first node.
delete_tail(): Remove the last node.
delete_node(node_ref): Given a direct reference to a node object, remove it from the list in $O(1)$ time.
reverse(): Reverse the list in-place (swap next and prev pointers for all nodes).display_forward(): Print from Head to Tail.
display_backward(): Print from Tail to Head.

```python
class Node:
    def __init__(self,val):
        self.val = val
        self.next = None
        self.prev = None

class DoublyLinkedList:
    def __init__(self):
        self.head = None

    # 1. Insert at Head
    def insert_at_head(self,val):
        new_node = Node(val)
        if not self.head:
            self.head = new_node
        else:
            new_node.next = self.head
            self.head.prev = new_node
            self.head = new_node

    # 2. Insert at Last (End)
    def append(self,val):
        new_node = Node(val)
        if not self.head:
            self.head = new_node
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node
            new_node.prev = current

    # 3. Insert in Between
    def insert_at(self,val,position):
        new_node = Node(val)
        if position == 0:
            self.insert_at_head(val)
        
        current = self.head
        count = 0
        while current and count < position - 1:
            current = current.next
            count += 1
        
        if current is None:
            print("Position out of bounds")
            return

        new_node.next = current.next
        new_node.prev = current
        if current.next:
            current.next.prev = new_node
        current.next = new_node
        

```

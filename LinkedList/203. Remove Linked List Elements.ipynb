````python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def removeElements(self, head: Optional[ListNode], val: int) -> Optional[ListNode]:
        if not head:
            return None                    
        
        curr = head
        while curr.next:
            if curr.next.val == val:
                curr.next = curr.next.next
            else:
                curr = curr.next
        if head.val == val:
            head = head.next
        return head
````

## **SOLUTION 1: Single Pass (Current Implementation)**

### **1.1 Intuition and Approach**

**Core Idea:**
- First pass through the list, skipping nodes with `curr.next.val == val`
- Check the head separately at the end (since we can't skip it in the normal loop)
- This avoids the need for a dummy node and handles removals in one traversal

**Why it works:**
- By checking `curr.next.val` in the loop, we can safely skip the next node if it matches
- Handles middle and tail nodes naturally
- Head check at the end catches all head removals

**Limitation:**
- The head node is checked last, after all other removals, which makes the logic slightly asymmetrical

### **1.2 Code Explanation in Detail**

```python
class Solution:
    def removeElements(self, head: Optional[ListNode], val: int) -> Optional[ListNode]:
        # Handle empty list
        if not head:
            return None
        
        # Traverse and remove non-head nodes
        curr = head
        while curr.next:
            if curr.next.val == val:
                curr.next = curr.next.next         # Skip next node
            else:
                curr = curr.next                   # Move forward
        
        # Handle head node removal (after all others)
        if head.val == val:
            head = head.next
        
        return head
```

**Line-by-Line Breakdown:**

| Line | Code | Explanation |
|------|------|-------------|
| 2-3 | `if not head: return None` | Edge case: empty list returns None |
| 6 | `curr = head` | Start pointer at head |
| 7 | `while curr.next:` | Traverse while next node exists |
| 8 | `if curr.next.val == val:` | Check if next node matches value to remove |
| 9 | `curr.next = curr.next.next` | Skip the matching node (pointer surgery) |
| 10-11 | `else: curr = curr.next` | Move forward if next doesn't match |
| 14-15 | Check head separately | After loop, check if head itself should be removed |
| 16 | `if head.val == val: head = head.next` | Remove head by pointing to next |
| 18 | `return head` | Return (possibly updated) head |

## **SOLUTION 2: Dummy Node Approach (More Elegant)**

### **2.1 Intuition and Approach**

**Core Idea:**
- Create a dummy node pointing to the head (handles head removal elegantly)
- Traverse from dummy, checking current node (not next node)
- All removals use uniform logic (no special head case)
- Return dummy.next as new head

**Why it's better:**
- Symmetric logic: all nodes treated the same way
- No special case for head node
- Cleaner, more readable code
- Easier to understand and debug

### **2.2 Code Explanation in Detail**

```python
class Solution:
    def removeElements(self, head: Optional[ListNode], val: int) -> Optional[ListNode]:
        # Create dummy node pointing to head
        dummy = ListNode(0)
        dummy.next = head
        
        # Traverse from dummy
        curr = dummy
        while curr.next:
            if curr.next.val == val:
                curr.next = curr.next.next     # Skip next node
            else:
                curr = curr.next               # Move forward
        
        return dummy.next                      # New head (skip dummy)
```

**Line-by-Line Breakdown:**

| Line | Code | Explanation |
|------|------|-------------|
| 2-3 | `dummy = ListNode(0)`, `dummy.next = head` | Create dummy node pointing to head |
| 6 | `curr = dummy` | Start from dummy (not head) |
| 7 | `while curr.next:` | Traverse while next exists |
| 8-9 | Skip if match | Same logic as Solution 1 |
| 10-11 | Move if no match | Same logic as Solution 1 |
| 13 | `return dummy.next` | Return new head (skip dummy) |

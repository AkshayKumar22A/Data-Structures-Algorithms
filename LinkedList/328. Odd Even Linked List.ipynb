Odd Even Linked List</a></h2><h3>Medium</h3><hr><p>Given the <code>head</code> of a singly linked list, group all the nodes with odd indices together followed by the nodes with even indices, and return <em>the reordered list</em>.</p>

<p>The <strong>first</strong> node is considered <strong>odd</strong>, and the <strong>second</strong> node is <strong>even</strong>, and so on.</p>

<p>Note that the relative order inside both the even and odd groups should remain as it was in the input.</p>

<p>You must solve the problem&nbsp;in <code>O(1)</code>&nbsp;extra space complexity and <code>O(n)</code> time complexity.</p>

````python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def oddEvenList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        odd = []
        even = []
        count = 0
        temp = head
        while temp:
            if count%2==0:
                odd.append(temp.val)
            elif count%2!=0:
                even.append(temp.val)
            temp = temp.next
            count+=1
        
        temp = head
        i = 0
        j = 0
        while temp:
            if i <= len(odd)-1:
                temp.val = odd[i]
                i+=1
            else:
                temp.val = even[j]
                j+=1
            temp = temp.next
        return head
````

## **SOLUTION 1: Array-Based Approach (Less Optimal)**

### **1.1 Intuition and Approach**

**Problem Statement:**
Rearrange a linked list so all nodes at odd indices (1st, 3rd, 5th...) appear before all nodes at even indices (2nd, 4th, 6th...).

**Approach:**
1. Use two arrays to store values: one for odd-indexed nodes, one for even-indexed nodes
2. Traverse the list once, collecting values at odd indices in one array and even indices in another
3. Traverse the list again, overwriting node values by first copying odd array values, then even array values
4. Return the modified list

**Why This Works:**
- Arrays allow easy separation of odd and even values
- Two passes ensure all values are collected and then reassigned in correct order

### **1.2 Code Explanation in Detail**

```python
class Solution:
    def oddEvenList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        odd = []          # Store values at odd indices (0, 2, 4, ...)
        even = []         # Store values at even indices (1, 3, 5, ...)
        count = 0
        temp = head
        
        # Pass 1: Collect values from list
        while temp:
            if count % 2 == 0:           # Even count = odd index (1st, 3rd, 5th)
                odd.append(temp.val)
            elif count % 2 != 0:         # Odd count = even index (2nd, 4th, 6th)
                even.append(temp.val)
            temp = temp.next
            count += 1
        
        # Pass 2: Reconstruct list with new order
        temp = head
        i = 0
        j = 0
        while temp:
            if i <= len(odd) - 1:        # First, assign all odd-indexed values
                temp.val = odd[i]
                i += 1
            else:                         # Then, assign all even-indexed values
                temp.val = even[j]
                j += 1
            temp = temp.next
        
        return head
```

### **1.5 Complexity Analysis**

**Time Complexity:**
- **Best/Average/Worst Case:** $O(n)$ - Two complete passes through the list
- Pass 1: $O(n)$ to collect values
- Pass 2: $O(n)$ to reconstruct
- Total: $O(n) + O(n) = O(n)$

**Space Complexity:**
- **Best/Average/Worst Case:** $O(n)$ - Need to store all node values in arrays
- `odd` array stores $\lceil n/2 \rceil$ values
- `even` array stores $\lfloor n/2 \rfloor$ values
- Total: $O(n)$ extra space ❌ **Violates O(1) constraint**

````python
from typing import Optional
# Definition for singly-linked list.
class ListNode:
     def __init__(self, val=0, next=None):
         self.val = val
         self.next = next

class Solution:
    def oddEvenList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        if head is None or head.next is None:
            return head
        odd = head
        even = head.next
        evenhead = head.next
        while odd and odd.next and even and even.next:
            odd.next = odd.next.next
            even.next = even.next.next
            odd = odd.next
            even = even.next
        odd.next = evenhead
        return head
````


## **SOLUTION 2: Optimal Pointer-Based Approach (RECOMMENDED)**

### **2.1 Intuition and Approach**

**Key Insight:**
Instead of storing values, manipulate pointers directly to separate odd and even indexed nodes into two chains, then connect them.

**Approach:**
1. Maintain two pointers: `odd` (for odd-indexed nodes) and `even` (for even-indexed nodes)
2. Keep `evenhead` pointing to the first even node to connect chains later
3. Traverse the list, alternately advancing both pointers by 2 steps:
   - `odd.next` skips one node (even node) and points to the next odd node
   - `even.next` skips one node (odd node) and points to the next even node
4. Connect the last odd node to the first even node
5. Return the head

**Why This Works:**
- Pointer manipulation preserves node structure while reordering
- No extra data structures needed
- Single pass through the list
- Meets O(1) space and O(n) time constraints ✓

### **2.2 Code Explanation in Detail**

```python
class Solution:
    def oddEvenList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        # Handle edge cases
        if head is None or head.next is None:
            return head
        
        # Initialize pointers
        odd = head                    # Points to 1st node (odd index)
        even = head.next              # Points to 2nd node (even index)
        evenhead = head.next          # Save reference to first even node
        
        # Separate odd and even chains
        while odd and odd.next and even and even.next:
            # Connect current odd node to next odd node (skip even node)
            odd.next = odd.next.next
            # Connect current even node to next even node (skip odd node)
            even.next = even.next.next
            # Move both pointers forward by 1 (effectively by 2 in original list)
            odd = odd.next
            even = even.next
        
        # Connect odd chain to even chain
        odd.next = evenhead
        return head
```

| Line | Code | Explanation |
|------|------|-------------|
| 2-3 | `if head is None or head.next is None:` | Handle empty/single node |
| 6-8 | Initialize `odd`, `even`, `evenhead` | Set up pointers for separation |

| 12 | `odd.next = odd.next.next` | Skip even node, connect to next odd |
| 14 | `even.next = even.next.next` | Skip odd node, connect to next even |
| 16-17 | `odd = odd.next`, `even = even.next` | Move forward in respective chains |
| 20 | `odd.next = evenhead` | Connect odd chain to even chain |
| 21 | `return head` | Return modified list |

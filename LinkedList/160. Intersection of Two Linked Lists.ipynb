<strong>Follow up:</strong> Could you write a solution that runs in <code>O(m + n)</code> time and use only <code>O(1)</code> memory?

````python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution:
    def getIntersectionNode(self, headA: ListNode, headB: ListNode) -> Optional[ListNode]:
        visited = set()
        temp = headA
        while temp:
            visited.add(temp)
            temp = temp.next
        temp = headB
        while temp:
            if temp in visited:
                return temp
            temp = temp.next
        return None
````

## **SOLUTION 1: Hash Set Approach (Extra Space)**

### **1.1 Intuition and Approach**

**Core Idea:**
- A node is an intersection if both pointers can point to the same object in memory.
- Store all node references from list A in a hash set.
- Traverse list B and check if any node exists in the set.
- The first match is the intersection point.

**Why it works:**
- Hash set membership check is O(1) on average.
- We leverage hash table to track visited nodes from list A.
- Works for any length combination and handles all cases.

### **1.2 Code Explanation in Detail**

```python
class Solution:
    def getIntersectionNode(self, headA: ListNode, headB: ListNode) -> Optional[ListNode]:
        visited = set()                  # Store node references from list A
        temp = headA
        
        # Pass 1: Add all nodes from list A to the set
        while temp:
            visited.add(temp)            # Add node object reference
            temp = temp.next
        
        # Pass 2: Check each node in list B
        temp = headB
        while temp:
            if temp in visited:          # Found intersection
                return temp
            temp = temp.next
        
        # No intersection found
        return None
```

| Line | Code | Explanation |
|------|------|-------------|
| 2 | `visited = set()` | Initialize empty hash set for node references |
| 3 | `temp = headA` | Start pointer at head of list A |
| 6 | `while temp:` | Traverse list A until None |
| 7 | `visited.add(temp)` | Add node object to set (stores memory reference) |
| 8 | `temp = temp.next` | Move to next node in list A |
| 11 | `temp = headB` | Switch to head of list B |
| 12 | `while temp:` | Traverse list B until None |
| 13 | `if temp in visited:` | Check if node exists in set (O(1) average) |
| 14 | `return temp` | Return the intersection node |
| 15 | `temp = temp.next` | Move to next node in list B |
| 18 | `return None` | No intersection found |

### **1.5 Time and Space Complexity**

**Time Complexity:**

| Case | Complexity | Explanation |
|------|-----------|-------------|
| **Best Case** | $O(m)$ | Intersection at head of B; traverse all of A + 1 node of B |
| **Average Case** | $O(m + n)$ | Traverse all of A + partial/full B |
| **Worst Case** | $O(m + n)$ | No intersection; traverse all of A and all of B |

**Overall:** $O(m + n)$ where m = length of A, n = length of B

**Analysis:**
- Pass 1: $O(m)$ — traverse all m nodes of list A and add to set
- Pass 2: $O(n)$ — traverse up to n nodes of list B with O(1) lookup each
- Total: $O(m) + O(n) = O(m + n)$
- Hash set operations (add, lookup) are O(1) on average

**Space Complexity:**

| Case | Complexity | Explanation |
|------|-----------|-------------|
| **Best Case** | $O(m)$ | Always store all nodes from A |
| **Average Case** | $O(m)$ | Store all m nodes from list A |
| **Worst Case** | $O(m)$ | Store all m nodes regardless of B |

**Overall:** $O(m)$ — Stores all m node references from list A ❌ **Does NOT meet follow-up O(1) requirement**

**Analysis:**
- Hash set stores up to m node references
- Does not depend on intersection position or list B length
- Each node reference takes constant space but total scales with A's length

````python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution:
    def getIntersectionNode(self, headA: ListNode, headB: ListNode) -> Optional[ListNode]:
        a = headA
        b = headB
        while a != b:
            if a is not None:
                a = a.next
            else:
                a = headB
            
            if b is not None:
                b = b.next
            else:
                b = headA
        return a
````


## **SOLUTION 2: Two-Pointer Switching (Optimal, O(1) Space)**

### **2.1 Intuition and Approach**

**Core Idea:**
- Use two pointers that start at the heads of respective lists.
- Each pointer advances one step at a time.
- When a pointer reaches the end (None), redirect it to the head of the OTHER list.
- If lists intersect, both pointers will eventually point to the same intersection node.
- If they don't intersect, both pointers will become None simultaneously.

**Why it works (Mathematical proof):**
- Let `m` = length of A, `n` = length of B, `c` = length of common suffix.
- Pointer A travels: A (m nodes) + B (n nodes) = m + n total steps
- Pointer B travels: B (n nodes) + A (m nodes) = m + n total steps
- If intersection exists at step k in the common part:
  - A reaches it after: (m - c) + k = m - c + k steps
  - B reaches it after: (n - c) + k = n - c + k steps
  - Both equal when scaled properly
- If no intersection: both pointers reach None after m + n steps

### **2.2 Code Explanation in Detail**

```python
class Solution:
    def getIntersectionNode(self, headA: ListNode, headB: ListNode) -> Optional[ListNode]:
        a = headA                        # Pointer 1 starts at A
        b = headB                        # Pointer 2 starts at B
        
        # Continue until pointers meet
        while a != b:
            # Advance pointer a; redirect to B when reaching end
            if a is not None:
                a = a.next
            else:
                a = headB
            
            # Advance pointer b; redirect to A when reaching end
            if b is not None:
                b = b.next
            else:
                b = headA
        
        # When loop exits, a == b (either intersection node or both None)
        return a
```

**Line-by-Line Breakdown:**

| Line | Code | Explanation |
|------|------|-------------|
| 2-3 | `a = headA`, `b = headB` | Initialize pointers at respective heads |
| 6 | `while a != b:` | Continue until pointers meet (object equality) |
| 8-9 | `if a is not None:` | Check if a is not at end |
| 9 | `a = a.next` | Advance a by one step |
| 10-11 | `else: a = headB` | a reached end, switch to B's head |
| 14-15 | `if b is not None:` | Check if b is not at end |
| 15 | `b = b.next` | Advance b by one step |
| 16-17 | `else: b = headA` | b reached end, switch to A's head |
| 20 | `return a` | Return intersection or None |

<h2><a href="https://leetcode.com/problems/linked-list-cycle-ii/">142. Linked List Cycle II</a></h2><h3>Medium</h3><hr><p>Given the <code>head</code> of a linked list, return <em>the node where the cycle begins. If there is no cycle, return </em><code>null</code>.</p>

<p>There is a cycle in a linked list if there is some node in the list that can be reached again by continuously following the <code>next</code> pointer. Internally, <code>pos</code> is used to denote the index of the node that tail&#39;s <code>next</code> pointer is connected to (<strong>0-indexed</strong>). It is <code>-1</code> if there is no cycle. <strong>Note that</strong> <code>pos</code> <strong>is not passed as a parameter</strong>.</p>

<p><strong>Do not modify</strong> the linked list.</p>

````python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution:
    def detectCycle(self, head: Optional[ListNode]) -> Optional[ListNode]:
        visited = set()
        temp = head
        while temp:
            visited.add(temp)
            temp = temp.next
            if temp in visited:
                return temp
        return None
        
````

## **APPROACH 1: Hash Set / Visited Nodes**

### **Intuition and Approach**

Simple and intuitive approach to find the cycle start:
1. **Track visited nodes**: Use a set to store node references as we traverse
2. **Find cycle start**: When we visit a node that's already in the set, that's the start of the cycle
3. **Return the node**: Return the node where the cycle begins, or None if no cycle

This approach is straightforward but uses O(n) extra space.

### **Code Explanation in Detail**

```python
class Solution:
    def detectCycle(self, head: Optional[ListNode]) -> Optional[ListNode]:
        visited = set()          # Set to store visited node references
        temp = head              # Pointer to traverse the list
        
        while temp:              # Continue while temp is not None
            visited.add(temp)    # Add current node to visited set
            temp = temp.next     # Move to next node
            if temp in visited:  # Check if we've seen this node before
                return temp      # Return the cycle start node
        
        return None              # No cycle found
```

**Line-by-line breakdown:**

- **Line 1**: `visited = set()` - Initialize empty set to store node references
- **Line 2**: `temp = head` - Start from the head node
- **Line 3**: `while temp:` - Loop continues while temp is not None
- **Line 4**: `visited.add(temp)` - Add current node to visited set
- **Line 5**: `temp = temp.next` - Move pointer to next node
- **Lines 6-7**: Check if next node is already visited → if yes, return it (cycle start)
- **Line 8**: Return None if loop ends without finding cycle

**Key difference from Problem 141**: We check `if temp in visited` **after** moving to the next node, so we return the first node we revisit (the cycle start).

### **Edge Cases**

1. **Empty list** (`head = None`):
   - `temp = None`
   - While loop: `while temp` → False
   - Returns `None` ✓

2. **Single node without cycle** (`head = [1]`):
   - Step 1: Add Node(1), temp = None
   - While loop exits
   - Returns `None` ✓

3. **Single node with self-cycle** (`head = [1]` where node points to itself):
   - Step 1: Add Node(1), temp = Node(1)
   - Check: Node(1) in visited → True
   - Returns `Node(1)` ✓

4. **Negative values** (`[-100, -50, 0, 50]` with cycle):
   - Node values don't affect detection
   - Returns the cycle start node correctly ✓

5. **Long list with late cycle** (10,000 nodes, cycle at position 5000):
   - Traverses nodes, detects cycle at position 5000
   - Returns the cycle start ✓

````python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution:
    def detectCycle(self, head: Optional[ListNode]) -> Optional[ListNode]:
        slow,fast = head,head
        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next
            if slow == fast:
                slow = head
                while slow != fast:
                    slow = slow.next
                    fast = fast.next
                return slow
        return None
````

## **APPROACH 2: Floyd's Cycle Detection with Math (Two Pointers) - OPTIMAL**

### **Intuition and Approach**

This is the **optimal O(1) space solution** using Floyd's algorithm with a mathematical insight:

**The Core Idea**:
1. **Detect cycle**: Use slow and fast pointers (like Problem 141) - if they meet, cycle exists
2. **Find start**: Once we detect a cycle, we use a mathematical property:
   - Distance from head to cycle start = Distance from collision point to cycle start
   - Reset slow pointer to head, move both pointers 1 step at a time
   - Where they meet again is the cycle start

**Mathematical Proof**:
Let's say:
- $a$ = distance from head to cycle start
- $b$ = distance from cycle start to collision point
- $c$ = remaining cycle length from collision point to cycle start

When fast and slow meet:
- Slow traveled: $a + b$ steps
- Fast traveled: $a + b + c + b = a + 2b + c$ steps (went around cycle once more)
- Fast is 2x speed: $2(a + b) = a + 2b + c$
- Solving: $2a + 2b = a + 2b + c$ → **$a = c$**

This means the distance from head to cycle start equals the distance from collision point to cycle start!

### **Code Explanation in Detail**

```python
class Solution:
    def detectCycle(self, head: Optional[ListNode]) -> Optional[ListNode]:
        slow, fast = head, head  # Initialize both pointers at head
        
        # Phase 1: Detect if cycle exists
        while fast and fast.next:
            slow = slow.next         # Move slow 1 step
            fast = fast.next.next    # Move fast 2 steps
            if slow == fast:         # Cycle detected
                slow = head          # Reset slow to head
                # Phase 2: Find cycle start
                while slow != fast:
                    slow = slow.next  # Move both 1 step at a time
                    fast = fast.next
                return slow          # Return cycle start
        
        return None                  # No cycle
```

**Line-by-line breakdown:**

- **Line 1**: `slow, fast = head, head` - Initialize both pointers at head
- **Lines 2-3**: While loop checks if fast pointer can move 2 steps safely
- **Line 4**: `slow = slow.next` - Move slow 1 step
- **Line 5**: `fast = fast.next.next` - Move fast 2 steps
- **Lines 6-7**: If they meet, cycle is confirmed
- **Line 8**: `slow = head` - Reset slow pointer to head for Phase 2
- **Lines 9-11**: Loop until slow and fast meet again - this point is the cycle start
- **Line 12**: Return the cycle start node
- **Line 14**: Return None if no cycle found

### **Dry Run in Detail**

**Example 1: List with cycle** - `[3, 2, 0, -4]` where pos = 1 (cycle starts at node 2)

```
Structure:  3 → 2 → 0 → -4
                ↑       ↓
                └───────┘

Distance a (head to cycle start):     1 (3 → 2)
Cycle nodes:                          2 → 0 → -4 → 2 (length = 3)
```

**Phase 1: Detect Cycle**

| Step | slow | fast | slow.val | fast.val | Meet? | Notes |
|------|---|---|---|---|---|---|
| Start | Node(3) | Node(3) | 3 | 3 | No | Both at head |
| 1 | Node(2) | Node(0) | 2 | 0 | No | slow moved 1, fast moved 2 |
| 2 | Node(0) | Node(2) | 0 | 2 | No | In cycle now |
| 3 | Node(-4) | Node(0) | -4 | 0 | No | - |
| 4 | Node(2) | Node(2) | 2 | 2 | **YES** | Collision! |

**Phase 2: Find Cycle Start**

After collision at Node(2):
- Reset slow to Node(3) (head)
- Fast remains at Node(2)

| Step | slow | fast | slow.val | fast.val | Meet? | Notes |
|------|---|---|---|---|---|---|
| Start | Node(3) | Node(2) | 3 | 2 | No | slow at head, fast at collision |
| 1 | Node(2) | Node(0) | 2 | 0 | No | - |
| 2 | Node(0) | Node(2) | 0 | 2 | No | - |
| 3 | Node(-4) | Node(0) | -4 | 0 | No | - |
| 4 | Node(2) | Node(2) | 2 | 2 | **YES** | Found cycle start! |

**Result**: Return Node(2) ✓

**Why this works mathematically**:
- Distance from head to cycle start (a) = 1
- From collision point Node(2) to cycle start Node(2) = 0 (it IS the start!)
- Actually more precisely: a = c (cycle length minus distance traveled in cycle to collision)

### **Edge Cases**

1. **Empty list** (`head = None`):
   - `slow = None, fast = None`
   - While: `while fast and fast.next` → False
   - Returns `None` ✓

2. **Single node without cycle** (`head = [1]`):
   - `slow = Node(1), fast = Node(1)`
   - While: `while fast and fast.next` → fast.next = None, so False
   - Returns `None` ✓

3. **Single node with self-cycle** (`head = [1]` where Node(1).next = Node(1)):
   - Phase 1, Step 1: slow = Node(1), fast = Node(1) → Meet!
   - Phase 2: slow = Node(1), fast = Node(1) → Already equal
   - Returns `Node(1)` ✓

4. **Negative values** (`[-5, -3, 0, 2]` with cycle):
   - Values don't affect pointer comparison logic
   - Finds cycle start correctly ✓

5. **Very long list** (10,000 nodes with cycle at position 100):
   - Eventually detects collision in cycle
   - Phase 2 finds the actual start
   - Returns start node ✓

6. **Cycle at the very beginning** (pos = 0):
   - Detected and returned correctly ✓

### **Time and Space Complexity**

| Metric | Best Case | Average Case | Worst Case |
|--------|-----------|--------------|-----------|
| **Time** | O(n) | O(n) | O(n) |
| **Space** | O(1) | O(1) | O(1) |

**Time Complexity: O(n)**
- Phase 1: Finding collision - slow travels at most n steps before collision/reaching end
- Phase 2: Finding cycle start - at most n more steps
- Total: O(n) + O(n) = O(n)
- **Best case O(n)**: Still need to traverse to find collision
- **Worst case O(n)**: No cycle requires full traversal

**Why Phase 2 takes at most n steps?**
- The property $a = c$ means slow and fast meet within n steps total

**Space Complexity: O(1)** ⭐
- Only two pointers (slow and fast) used
- No extra data structures
- **Answers the follow-up question: "Can you solve it using O(1) space?"**

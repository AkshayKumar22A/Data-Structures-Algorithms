# Stack Implementation in Python

## What is a Stack?

A **Stack** is a linear data structure that follows the **LIFO (Last In First Out)** principle. The last element inserted is the first one to be removed, similar to a stack of plates where you add/remove from the top.

### Visual Representation

```
Push(5) →  Push(10) →  Push(15) →  Push(5)
   ↓         ↓            ↓          ↓
  [5]       [10]         [15]       [5]  ← Top (Next to pop)
             [5]         [10]       [15]
                          [5]       [10]
                                    [5]   ← Bottom
```

---

## Implementation 1: Stack Using List (Array-Based)

### Intuition and Approach

- Use a **Python list** to store stack elements
- Elements are added/removed from the **end** of the list (rear)
- The end of the list represents the **top** of the stack
- All operations (`push`, `pop`, `top`) access only the last element, making them optimal

### Structure

```
List Index:  [0]  [1]  [2]  [3]
Elements:    [5] [10] [15] [5]  ← items[-1] is the top
             ↑                   ↑
          Bottom              Top (push/pop here)
```

### Code Explanation (Line-by-Line)

```python
class Stack:
    def __init__(self):
        self.items = []
```
- Initialize an empty list `items` to store stack elements
- `__init__` is called when creating a new Stack instance

```python
    def is_empty(self):
        return len(self.items) == 0
```
- Check if stack has no elements by verifying list length is 0
- Used to prevent `pop()` or `top()` on empty stack
- Time: O(1) because `len()` is a built-in that runs in constant time

```python
    def push(self, item):
        self.items.append(item)
```
- Add element `item` to the **end** of the list using `append()`
- `append()` is O(1) amortized because list grows dynamically
- The new element becomes the **top** of the stack

```python
    def pop(self):
        if len(self.items) == 0:
            return "Cannot pop, stack is empty"
        
        x = self.items.pop()
        return x
```
- First, check if stack is empty to avoid IndexError
- `self.items.pop()` removes and returns the **last element** (LIFO)
- Returns the popped value to the caller
- Why two lines (x = ...; return x)? For clarity, though could be written as `return self.items.pop()`

```python
    def top(self):
        if len(self.items) == 0:
            return "Cannot top, stack is empty"
        return self.items[-1]
```
- Return the **last element** without removing it (`self.items[-1]`)
- Negative index -1 means "last element" in Python
- Useful for peeking at what will be popped next

```python
    def size(self):
        return len(self.items)
```
- Return the count of elements currently in stack
- Let caller know how many elements exist

```python
    def display(self):
        return self.items
```
- Return the entire list for debugging/visualization
- Shows all elements from oldest (index 0) to newest (index -1)


### Dry Run: Step-by-Step Execution

```
Initial: stack = Stack()  →  items = []

Step 1: stack.push(5)
        items = [5]
        Top = 5

Step 2: stack.push(10)
        items = [5, 10]
        Top = 10

Step 3: stack.push(15)
        items = [5, 10, 15]
        Top = 15

Step 4: stack.push(5)
        items = [5, 10, 15, 5]
        Top = 5

Step 5: stack.display()
        Output: [5, 10, 15, 5]

Step 6: stack.pop()
        Remove items[-1] = 5
        items = [5, 10, 15]
        Returns: 5

Step 7: stack.display()
        Output: [5, 10, 15]

Step 8: stack.top()
        Peek items[-1] = 15
        items = [5, 10, 15]  (unchanged)
        Returns: 15

Step 9: stack.is_empty()
        len(items) == 0?  False
        Returns: False
```

## Time & Space Complexity: List-Based Stack

| Operation | Best Case | Average Case | Worst Case | Explanation |
|-----------|-----------|--------------|-----------|-------------|
| `push`    | O(1) | O(1) amortized | O(n)* | Append to list usually O(1), rarely O(n) when list capacity exceeded |
| `pop`     | O(1) | O(1) | O(1) | Remove from end is always constant |
| `top`     | O(1) | O(1) | O(1) | Access by index [-1] is constant |
| `is_empty` | O(1) | O(1) | O(1) | Length check is constant |
| `size`    | O(1) | O(1) | O(1) | Length retrieval is constant |

**Space Complexity:** O(n) where n is the number of elements stored

*Note: `append()` is O(1) **amortized** because Python lists grow exponentially (doubling capacity), so cost averages to O(1) per operation over time.

## Edge Cases: List-Based Stack

### Case 1: Empty Stack Operations
```python
stack = Stack()
print(stack.pop())      # "Cannot pop, stack is empty"
print(stack.top())      # "Cannot top, stack is empty"
print(stack.is_empty()) # True
print(stack.size())     # 0
```

### Case 2: Single Element
```python
stack = Stack()
stack.push(10)
print(stack.top())    # 10
print(stack.pop())    # 10 (removed)
print(stack.size())   # 0
```

### Case 3: Negative Numbers
```python
stack = Stack()
stack.push(-5)
stack.push(-100)
print(stack.pop())    # -100 (negative handled fine)
print(stack.top())    # -5
```

### Case 4: Zero Value
```python
stack = Stack()
stack.push(0)
if stack.pop():  # ✗ WARNING: 0 is falsy!
    print("Not empty")
else:
    print("0 popped")  # This runs even though we had an element
```
When checking if pop was successful, don't use truthiness; check against `None`.

### Case 5: Mixed Data Types
```python
stack = Stack()
stack.push(5)
stack.push("hello")
stack.push([1, 2, 3])
print(stack.pop())    # [1, 2, 3] - works fine!
```
**Note:** Stack stores any Python object, not just integers.

---

````python
class Stack:
    def __init__(self):
        self.items = []
    
    def is_empty(self):
        return len(self.items) == 0
    
    def push(self,item):
        self.items.append(item)
    
    def pop(self):
        if len(self.items) == 0:
            return "Cannot pop, stack is empty"
        
        x = self.items.pop()
        return x
    
    def top(self):
        if len(self.items) == 0:
            return "Cannot top, stack is empty"
        return self.items[-1]
    
    def size(self):
        return len(self.items)

    def display(self):
        return self.items

stack = Stack()
stack.push(5)
stack.push(10)
stack.push(15)
stack.push(5)

print(f"Stack content = {stack.display()}")
print(f"Popped item = {stack.pop()}")
print(f"Stack content = {stack.display()}")
print(f"Top item after pop = {stack.top()}")
print(f"Stack id empty = {stack.is_empty()}")

````

## Implementation 2: Stack Using Doubly Linked List (DLL-Based)

### Intuition and Approach

- Use a **doubly linked list** where each node has:
  - `val`: the data
  - `next`: pointer to next node
  - `prev`: pointer to previous node
- Maintain `head` (oldest element) and `tail` (newest element/top)
- `push` adds node at tail (top)
- `pop` removes node from tail (top)
- Main advantage: **no fixed storage pre-allocation**; memory grows dynamically

### Structure

```
Node: [prev] ← [val] → [next]

Example after push(5), push(10), push(15):

Head                           Tail
 ↓                             ↓
[5] ↔ [10] ↔ [15]
      ← Doubly linked →
```

### Code Explanation: Node Class

```python
class Node:
    def __init__(self, val):
        self.val = val
        self.next = None
        self.prev = None
```
- Each node stores one element
- `next` points to the node added after (closer to tail)
- `prev` points to the node added before (closer to head)
- Initially both pointers are `None`

### Code Explanation: Stack_DLL Class

```python
class Stack_DLL:
    def __init__(self):
        self.head = None
        self.tail = None
        self.count = 0
```
- `head`: points to the oldest node (bottom of stack)
- `tail`: points to the newest node (top of stack)
- `count`: tracks number of elements (for size() method)
- All initialized to None/0 for empty stack

```python
    def is_empty(self):
        return self.head is None
```
- If head is None, the entire list is empty
- Could also check `self.count == 0` or `self.tail is None`

```python
    def push(self, item):
        newNode = Node(item)
        if self.head is None and self.tail is None:
            self.head = newNode
            self.tail = newNode
        else:
            newNode.prev = self.tail
            self.tail.next = newNode
            self.tail = newNode
        self.count += 1
```

**Case 1: Empty Stack**
- Create new node with `item`
- Set both head and tail to this node (it's the only node)

**Case 2: Stack Has Elements**
- Create new node with `item`
- Link new node: set its `prev` to current tail
- Link current tail: set its `next` to new node
- Move tail pointer to new node (it's now the top)
- Increment count

**Why this structure?** Allows O(1) push at top AND O(1) pop from top

```python
    def pop(self):
        if self.is_empty():
            print("Cannot pop, stack is empty")
            return None
        
        elif self.tail == self.head:
            self.head = None
        
        temp = self.tail

        if self.head == self.tail:
            self.head = None
            self.tail = None
        else:
            self.tail = self.tail.prev
            self.tail.next = None
        
        self.count -= 1
        return temp.val
```

**Logic Breakdown:**
1. Check if stack is empty; return None if it is
2. Save reference to tail node in `temp`
3. Handle popping the last element: set both head and tail to None
4. For other cases: move tail to previous node and remove its `next` link
5. Decrement count and return the popped value

**Why `temp = self.tail` first?** We need to return the value before removing the node

```python
    def top(self):
        if self.is_empty():
            print("Cannot top, stack is empty")
            return
        return self.tail.val
```
- Return the value of the tail node without removing it
- `None` return if empty

```python
    def size(self):
        return self.count
```
- O(1) because we track count; no need to traverse the list

```python
    def display(self):
        print("From head to tail:")
        temp = self.head
        while temp:
            print(temp.val, end=" -> ")
            temp = temp.next

        print("From tail to head:")
        temp = self.tail
        while temp:
            print(temp.val, end=" -> ")
            temp = temp.prev
        print("None")
```
- Show elements from oldest to newest (head → tail)
- Show elements from newest to oldest (tail → head)
- Validates that both directions link correctly (doubly-linked integrity check)

```

        tail = Node(10)
        return tail.val = 10

```

## Time & Space Complexity: DLL-Based Stack

| Operation | Best Case | Average Case | Worst Case | Explanation |
|-----------|-----------|--------------|-----------|-------------|
| `push`    | O(1) | O(1) | O(1) | Create node and link at tail; constant operations |
| `pop`     | O(1) | O(1) | O(1) | Unlink tail node; constant operations |
| `top`     | O(1) | O(1) | O(1) | Access tail; no traversal needed |
| `is_empty` | O(1) | O(1) | O(1) | Check head pointer |
| `size`    | O(1) | O(1) | O(1) | We track count variable |
| `display` | O(n) | O(n) | O(n) | Must traverse all n nodes |

**Space Complexity:** O(n) + O(n) overhead for pointers
- Each node needs extra memory for `prev` and `next` pointers
- Slightly more memory than list-based stack

## Edge Cases: DLL Stack

### Case 1: Empty Stack
```python
s = Stack_DLL()
s.pop()      # "Cannot pop, stack is empty" + None
s.top()      # "Cannot top, stack is empty" + None
s.is_empty() # True
```

### Case 2: Single Element (Important!)
```python

s.push(5)

# After pop, head=None and tail=None
s.is_empty()    # True
```
**Code handles this:** The condition `if self.head == self.tail` catches this case

### Case 3: Negative Numbers & Zero
```python
s = Stack_DLL()
s.push(-50)
s.push(0)
print(s.pop())     # 0 (handled correctly)
print(s.pop())     # -50
```

### Case 4: Memory Leak Risk (Important!)
```python
# ✗ Potential issue: if popped node is referenced elsewhere,
#   it's not truly freed in Python (GC handles it)
popped_node = s.pop()  # Returns int value, not node
```
**In Python:** Garbage collector automatically frees unreferenced nodes

## Comparison: List vs DLL Stack

| Feature | List-Based | DLL-Based |
|---------|-----------|-----------|
| **Implementation Complexity** | Simple | Complex |
| **Space per Element** | Minimal | Extra (2 pointers) |
| **Cache Locality** | Good (contiguous) | Poor (scattered) |
| **Dynamic Resizing** | O(n) rarely | O(1) always |
| **Learning Value** | Foundation | Pointer practice |
| **Real-World Use** | Preferred | Educational |

---


## Conclusion

Both implementations achieve the **LIFO stack behavior** with **O(1) push/pop operations**:
- **List-based:** Simpler, more practical, better performance
- **DLL-based:** More educational, showcases pointer manipulation, good interview material

````python
class Node:
    def __init__(self,val):
        self.val = val
        self.next = None
        self.prev = None

class Stack_DLL:
    def __init__(self):
        self.head = None
        self.tail = None
        self.count = 0
    
    def is_empty(self):
        return self.head is None
    
    def push(self,item):
        newNode = Node(item)
        if self.head is None and self.tail is None:
            self.head = newNode
            self.tail = newNode
        else:
            newNode.prev = self.tail
            self.tail.next = newNode
            self.tail = newNode
        self.count += 1
    
    def pop(self):
        if self.is_empty():
            print("Cannot pop, stack is empty")
            return None
        
        elif self.tail == self.head:
            self.head = None
        
        temp = self.tail

        if self.head == self.tail:
            self.head = None
            self.tail = None
        else:
            self.tail = self.tail.prev
            self.tail.next = None
        
        self.count -= 1
        return temp.val
        

    
    def top(self):
        if self.is_empty():
            print ("Cannot top, stack is empty")
            return
        return self.tail.val
    
    def size(self):
        return self.count
    
    def display(self):
        print("From head to tail:")
        temp = self.head
        while temp:
            print(temp.val, end=" -> ")
            temp = temp.next

        print("From tail to head:")
        temp = self.tail
        while temp:
            print(temp.val, end=" -> ")
            temp = temp.prev
        print("None")

if __name__ == '__main__':
    s = Stack_DLL()
    s.push(5)
    s.push(10)
    s.push(15)
    s.push(20)
    print(s.pop())
    #print(s.pop())
    #print(s.pop())
    #print(s.pop())
    print(s.pop())

    print(s.top())
    print(s.size())
    s.display()
    print(s.is_empty())

````

### Intuition and Approach

A stack is **last‑in‑first‑out (LIFO)**. A queue is **first‑in‑first‑out (FIFO)**. To simulate a stack with a queue we must somehow make the *most recently pushed element* appear at the **front** of the queue.

The trick used here is simple:
1. Always push new items to the back of a single queue.
2. Immediately after each push, rotate the queue by repeatedly removing the front element and appending it to the back. Do this `len(queue)-1` times.

After the rotation the newest element occupies the front of the queue, which allows pop/top operations to behave like a stack.

---

### Code Explanation

```python
from collections import deque
class MyStack:

    def __init__(self):
        self.queue = deque()
```
* We use `collections.deque` as a double‑ended queue. Only standard queue operations (`append`/`popleft`, `len`) are used.

```python
    def push(self, x: int) -> None:
        self.queue.append(x)
        for _ in range(len(self.queue)-1):
            self.queue.append(self.queue.popleft())
```
* Push the value `x` to the back of the queue (`append`).
* The `for` loop runs exactly `n-1` times when the queue length is `n` after appending.
* Each iteration removes the current front element (`popleft`) and re‑appends it to the back. This effectively rotates the queue, moving `x` from the rear to the front.

```python
    def pop(self) -> int:
        if len(self.queue) == 0:
            return
        return self.queue.popleft()
```
* `pop` simply removes and returns the front element. If the queue is empty we return `None` (consistent with original stub), though in practice all calls are guaranteed valid.

```python
    def top(self) -> int:
        if len(self.queue) == 0:
            return
        return self.queue[0]
```
* `top` peeks at the front without removing it.

```python
    def empty(self) -> bool:
        return len(self.queue) == 0
```
* `empty` checks whether the queue holds any items.


Each method uses only allowed queue operations.

### Dry Run (example sequence)


Operations: push(1), push(2), push(3), top(), pop(), empty()

1. `push(1)`
   * queue before: `[]` → after `append(1)`: `[1]`
   * rotation loop runs 0 times → queue remains `[1]`
2. `push(2)`
   * append → `[1,2]`
   * rotate once: remove `1`, append → `[2,1]`
3. `push(3)`
   * append → `[2,1,3]`
   * rotate twice:
     * remove `2`, append → `[1,3,2]`
     * remove `1`, append → `[3,2,1]`
   * Now front is `3`, the most recent element.
4. `top()` → peek front → returns `3`.
5. `pop()` → remove front → returns `3`, queue becomes `[2,1]`.
6. `empty()` → `len(queue) == 0` ? No → returns `False`.

This matches stack behaviour.

### Edge Cases & Notes

* **Empty stack**: `pop`/`top` return `None`. LeetCode guarantees these calls are valid so they won’t be called on an empty stack, but the guard protects against misuse.
* **Zero or negative values**: allowed because values are simply stored; the algorithm is value‑agnostic.
* **Multiple pushes**: rotation cost accumulates; see complexity below.

### Time & Space Complexity

| Operation | Best | Average | Worst | Explanation |
|-----------|------|---------|-------|-------------|
| `push`    | **O(1)** | **O(n)** | **O(n)** | Must rotate the queue (`n` = queue size after push). Best-case when queue is empty or small. Average/worst linear. |
| `pop`     | **O(1)** | **O(1)** | **O(1)** | Single `popleft`. |
| `top`     | **O(1)** | **O(1)** | **O(1)** | Index access at front. |
| `empty`   | **O(1)** | **O(1)** | **O(1)** | Length check. |

* **Space**: O(n) additional space for the deque where `n` is number of elements stored. No auxiliary data structures.

In practice, `push` dominates the cost because of the rotations, but overall each element undergoes at most one rotation per push.

---

````python
from collections import deque
class MyStack:

    def __init__(self):
        self.queue = deque()
        

    def push(self, x: int) -> None:
        self.queue.append(x)
        for _ in range(len(self.queue)-1):
            self.queue.append(self.queue.popleft())
        

    def pop(self) -> int:
        if len(self.queue) == 0:
            return
        return self.queue.popleft()

    def top(self) -> int:
        if len(self.queue) == 0:
            return
        return self.queue[0]

    def empty(self) -> bool:
        return len(self.queue) == 0
        


obj = MyStack()
obj.push(5)
param_2 = obj.pop()
param_3 = obj.top()
param_4 = obj.empty()
````

### Alternative Implementation (Two Queues or Pop-Costly)

Another valid design uses **two queues** and makes the **pop** operation the expensive one instead of `push`. This pattern still respects the constraint of using only queue operations.

**Intuition:**
- Maintain two queues, `q1` holding the current stack elements and `q2` as a temporary buffer.
- To push, just enqueue to `q1` (O(1)).
- To pop, move every element except the last from `q1` into `q2`. The last element left in `q1` is the top of the stack; pop and return it. Afterwards, swap the roles of `q1` and `q2` so that `q1` again represents the stack.

```python
from collections import deque

class MyStackAlt:
    def __init__(self):
        self.q1 = deque()
        self.q2 = deque()

    def push(self, x: int) -> None:
        self.q1.append(x)             # enqueue in primary queue

    def pop(self) -> int:
        # move all but last item from q1 to q2
        while len(self.q1) > 1:
            self.q2.append(self.q1.popleft())
        # last element in q1 is the "top"
        res = self.q1.popleft()
        # swap names so q1 always holds the stack
        self.q1, self.q2 = self.q2, self.q1
        return res

    def top(self) -> int:
        return self.q1[-1]  # look at the most recently pushed element

    def empty(self) -> bool:
        return not self.q1
```

**Code Explanation:**
- `__init__`: two empty deques.
- `push`: straightforward enqueue; no rotations needed.
- `pop`: the loop moves `n-1` elements, preserving their order, leaving only the newest element which behaves as the stack `top`. After retrieving it, we swap the queues to avoid copying elements back.
- `top` and `empty` are constant-time operations accessing `q1`.

**Dry Run:**
```
ops: push(1), push(2), push(3), pop(), top(), empty()
```
1. `push(1)` → q1=[1], q2=[]
2. `push(2)` → q1=[1,2]
3. `push(3)` → q1=[1,2,3]
4. `pop()` → move 1→q2 ([1]), 2→q2 ([1,2]); q1 now [3], pop 3, swap → q1=[1,2], q2=[]; return 3
5. `top()` → q1[-1] = 2
6. `empty()` → False

**Edge Cases:**
- Empty structure: `pop`/`top` assume valid calls; if `q1` empty they would raise `IndexError` or could return `None` if guarded.
- Values zero or negative unaffected.
- Performance unaffected by value magnitude.

**Complexity:**
| Operation | Best | Average | Worst | Notes |
|-----------|------|---------|-------|-------|
| `push`    | O(1) | O(1)    | O(1)  | simple enqueue |
| `pop`     | O(1) | O(n)    | O(n)  | loop moves elements |
| `top`     | O(1) | O(1)    | O(1)  | peek rear of q1 |
| `empty`   | O(1) | O(1)    | O(1)  | length check |

*Space*: O(n) for the two queues combined.

This alternative further illustrates the flexibility in trading costs between operations when simulating one ADT with another.

Feel free to instantiate and test `MyStackAlt` alongside the original `MyStack` to compare their behaviors and performance.

````python
from collections import deque

class MyStackAlt:
    def __init__(self):
        self.q1 = deque()
        self.q2 = deque()

    def push(self, x: int) -> None:
        self.q1.append(x)      

    def pop(self) -> int:
        while len(self.q1) > 1:
            self.q2.append(self.q1.popleft())
        res = self.q1.popleft()
        self.q1, self.q2 = self.q2, self.q1
        return res

    def top(self) -> int:
        return self.q1[-1] 

    def empty(self) -> bool:
        return not self.q1
````

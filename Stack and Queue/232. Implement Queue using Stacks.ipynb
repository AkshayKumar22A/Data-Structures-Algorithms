### 🧠 Intuition and Approach

A queue is **first‑in‑first‑out (FIFO)**. A stack is **last‑in‑first‑out (LIFO)**. To simulate a queue using stacks, we need to reverse the natural LIFO behavior of stacks.

The strategy is:
1. Use two stacks: `st1` (main stack) and `st2` (auxiliary/temporary stack).
2. On every `push`, first empty `st1` by popping its elements into `st2` (reversing order), then push the new element onto the now-empty `st1`, and finally pop all elements from `st2` back onto `st1`.
3. After this process, the newest element ends up at the **bottom** of `st1` and will be the **last to be popped**, ensuring FIFO behavior.
4. Operations `pop`, `peek`, and `empty` work directly on `st1`.

---

### 💡 Code Explanation

```python
from collections import deque

class MyQueue:

    def __init__(self):
        self.st1 = deque()  # main stack storing queue elements
        self.st2 = deque()  # auxiliary/temporary stack
```
* We use `deque` (double-ended queue) to simulate stacks. Only `append` (push) and `pop` are used for stack operations.

```python
    def push(self, x: int) -> None:
        while self.st1:
            self.st2.append(self.st1.pop())
```
* Empty `st1` entirely by popping each element and pushing it onto `st2`.
* This reverses the order of elements in `st1`: if `st1` was `[a, b, c]` (c on top), it becomes `[c, b, a]` in `st2` (a on top after the loop).

```python
        self.st1.append(x)
```
* Push the new element `x` onto the now-empty `st1`. It sits at the bottom (to be removed last).

```python
        while self.st2:
            self.st1.append(self.st2.pop())
```
* Pop all elements from `st2` back onto `st1` in reverse, restoring the original order but with the new element now at the bottom.
* This ensures FIFO: oldest elements remain on top and will be popped first.

```python
    def pop(self) -> int:
        if not self.st1:
            return -1
        top_element = self.st1.pop()
        return top_element
```
* `pop` removes and returns the top element of `st1`, which is the **front** of the queue (the oldest element).
* If the queue is empty, return `-1` (guard against invalid calls, though the problem guarantees valid calls).

```python
    def peek(self) -> int:
        if not self.st1:
            return -1
        return self.st1[-1]
```
* `peek` returns the top element of `st1` **without removing it**, giving the front of the queue.

```python
    def empty(self) -> bool:
        return not self.st1
```
* `empty` returns `True` if `st1` is empty, `False` otherwise.

### 🧪 Dry Run

```
Operations: push(1), push(2), push(3), peek(), pop(), pop(), empty()
```

**Initial state:** st1=[], st2=[]

1. **push(1)**
   * st1 is empty → while loop doesn't execute.
   * st1.append(1) → st1=[1]
   * st2 is empty → second while loop doesn't execute.
   * **Result:** st1=[1], st2=[]

2. **push(2)**
   * Move st1 to st2: pop 1 from st1, push to st2 → st1=[], st2=[1]
   * st1.append(2) → st1=[2]
   * Move st2 back to st1: pop 1 from st2, push to st1 → st1=[2,1]
   * **Result:** st1=[2,1], st2=[] (2 is on top, 1 is below—FIFO ready: 1 will be popped first)

3. **push(3)**
   * Move st1 to st2: pop 1 then pop 2 from st1 → st1=[], st2=[1,2]
   * st1.append(3) → st1=[3]
   * Move st2 back to st1: pop 2, pop 1 → st1=[3,2,1]
   * **Result:** st1=[3,2,1], st2=[] (front=1, will pop in order 1→2→3)

4. **peek()**
   * return st1[-1] = st1[0] = 1 (front of queue)

5. **pop()**
   * not st1 is False
   * pop from st1 → returns 1, st1=[3,2]

6. **pop()**
   * pop from st1 → returns 2, st1=[3]

7. **empty()**
   * st1 has one element (3) → not st1 is False → return False

**Expected Output:** [null, null, null, 1, 1, 2, false] ✓

### ⚠️ Edge Cases & Notes

* **Empty queue:** Calling `pop()` or `peek()` on an empty queue returns `-1`. The problem guarantees valid calls, so this is defensive.
* **Zero value:** Storing 0 is no different from any other value; the algorithm is value-agnostic.
* **Negative numbers:** Allowed within the constraint range; treated like any other value.
* **Single element:** `push(x)` followed by `pop()` works correctly; the two while loops handle it gracefully.
* **Repeated operations:** Multiple pushes and pops maintain the FIFO property throughout.

---

### ⏱ Time & 📦 Space Complexity

| Operation | Best Case | Average Case | Worst Case | Explanation |
|-----------|-----------|--------------|-----------|-------------|
| `push`    | **O(n)** | **O(n)** | **O(n)** | Always moves all n elements from st1 → st2 → back to st1 (2n transfers). |
| `pop`     | **O(1)** | **O(1)** | **O(1)** | Single pop from top of st1. |
| `peek`    | **O(1)** | **O(1)** | **O(1)** | Constant-time access to top element. |
| `empty`   | **O(1)** | **O(1)** | **O(1)** | Length/emptiness check. |

**Space Complexity:** O(n) for storing n elements across the two stacks combined.

**Note on Follow‑up (Amortized O(1) for each operation):**
The problem asks: "Can you implement the queue such that each operation is amortized O(1)?"
This current approach does **not** achieve amortized O(1) for push because every push costs O(n).  
An **optimized approach** uses two stacks differently:
- `st1` for pushes (front of queue).
- `st2` for pops (rear of queue).
- Only when we need to pop and `st2` is empty do we reverse st1 into st2 (**amortized** O(1)).
This way, most operations run in O(1), with occasional O(n) operations that average out.

---


````python
from collections import deque
class MyQueue:

    def __init__(self):
        self.st1 = deque()
        self.st2 = deque()
        

    def push(self, x: int) -> None:
        while self.st1:
            self.st2.append(self.st1.pop())
        self.st1.append(x)
        while self.st2:
            self.st1.append(self.st2.pop())
        

    def pop(self) -> int:
        if not self.st1:
            return -1
        top_element = self.st1.pop()
        return top_element

    def peek(self) -> int:
        if not self.st1:
            return -1
        return self.st1[-1]

    def empty(self) -> bool:
        return not self.st1
        


# Your MyQueue object will be instantiated and called as such:
obj = MyQueue()
obj.push(5)
param_2 = obj.pop()
param_3 = obj.peek()
param_4 = obj.empty()
````

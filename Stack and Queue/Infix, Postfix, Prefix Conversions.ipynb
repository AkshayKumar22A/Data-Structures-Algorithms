# Expression Notations: Infix, Postfix, Prefix

## Components

### Operators
- Arithmetic: `+`, `-`, `*`, `/`
- Exponentiation: `^` (highest precedence)
- Parentheses: `(`, `)` (for grouping)

### Operands
- Letters: `a-z`, `A-Z`
- Digits: `0-9`
- Each character is treated as a separate operand

### Operator Precedence Hierarchy
```
Highest: ^ (right-associative)
Medium:  * , /
Lowest:  + , -
```

## Expression Types

### 1. Infix Notation
- **Format**: `operand operator operand`
- **Example**: `a + b * c`
- **Usage**: Most common in programming and mathematics
- **Evaluation**: Requires precedence rules and parentheses

### 2. Postfix Notation (Reverse Polish Notation)
- **Format**: `operand operand operator`
- **Example**: `a b c * +`
- **Usage**: Used in stack-based calculators, compilers
- **Evaluation**: Simple left-to-right processing with stack

### 3. Prefix Notation (Polish Notation)
- **Format**: `operator operand operand`
- **Example**: `+ a * b c`
- **Usage**: Used in Lisp programming, expression trees
- **Evaluation**: Right-to-left processing

## Conversion Relationships

```
Infix ↔ Postfix ↔ Prefix
  ↓       ↓       ↓
Human  Stack   Tree
Input  Eval   Parse
```

# Infix to Postfix Conversion

## Problem Statement

**Given a string `s` representing an infix expression, convert this infix expression to a postfix expression.**

### Expression Types
- **Infix**: Operators are between operands (e.g., `a + b`)
- **Postfix**: Operators follow operands (e.g., `a b +`)
- **Prefix**: Operators precede operands (e.g., `+ a b`)

### Operator Precedence & Associativity
| Operator | Precedence | Associativity |
|----------|------------|---------------|
| `^`      | Highest    | Right to Left |
| `*`, `/` | Medium     | Left to Right  |
| `+`, `-` | Lowest     | Left to Right  |

### Examples

**Example 1:**
```
Input: s = "a*(b+c)/d"
Output: "abc+*d/"
Explanation: a*(b+c)/d → a*(bc+)/d → a*(bc+*) → abc+*d/
```

**Example 2:**
```
Input: s = "a+b*c+d"
Output: "abc*+d+"
Explanation: a+b*c+d → a+b*c+d → a+bc*+d → abc*+d+
```

**Example 3:**
```
Input: s = "(a+b)*(c+d)"
Output: "ab+cd+*"
Explanation: (a+b)*(c+d) → ab+ * cd+ → ab+cd+*
```

### Constraints
- `1 ≤ s.length ≤ 5*10³`
- `s[i]` can be operand (a-z, A-Z, 0-9), operator (+, -, *, /, ^), or parenthesis ((, ))

### Intuition and Approach
Infix expressions place operators between operands (e.g., `a + b * c`), making evaluation complex due to precedence and associativity. Converting to postfix (e.g., `a b c * +`) simplifies evaluation by placing operators after operands. The approach uses a stack for operators: scan left to right, append operands to result, push operators to stack, and pop based on precedence rules.

### Code Explanation
- `precedence(ch)`: Returns precedence (1 for +-, 2 for */, 3 for ^, 0 else). This ensures correct operator ordering.
- `infixtoPostfix(s)`: Initializes result list and stack. For each char: if operand, append to result; if '(', push to stack; if ')', pop until '('; else, pop higher/equal precedence operators, then push current.
- After loop, pop remaining operators. Join result into string.

### Dry Run
Input: `a*(b+c)/d`
- 'a': append 'a' → result=['a']
- '*': push '*' → stack=['*']
- '(': push '(' → stack=['*','(']
- 'b': append 'b' → result=['a','b']
- '+': push '+' → stack=['*','(','+']
- 'c': append 'c' → result=['a','b','c']
- ')': pop '+' → result=['a','b','c','+'], pop '(' → stack=['*']
- '/': precedence('/')=2 > stack.top('*')=2? No, but since '*' has same, and left assoc, pop '*' → result=['a','b','c','+','*'], push '/' → stack=['/']
- 'd': append 'd' → result=['a','b','c','+','*','d']
- End: pop '/' → result=['a','b','c','+','*','d','/'] → "abc+*d/"

### Potential Edge Cases
- Simple: "a+b" → "ab+"
- No operators: "abc" → "abc"
- Parentheses: "(a+b)*c" → "ab+c*"
- Precedence: "a+b*c" → "abc*+"
- Empty/invalid: Not handled, assumes valid.

### Time and Space Complexity
- Time: O(N) - linear scan.
- Space: O(N) - stack and result.
- Best/Worst/Average: O(N).

````python
# Infix to postfix
class Solution:
    def precedence(self,ch):
        if ch == "+" or ch == "-":
            return 1
        if ch == "*" or ch == "/":
            return 2
        if ch == "^":
            return 3
        return 0

    def infixtoPostfix(self, s):
        result = []
        stack = []

        for char in s:
            if ("a" <= char <= "z") or ("A" <= char <= "Z") or ("0" <= char <= "9"):
                result.append(char)
            elif char == "(":
                stack.append(char)
            elif char == ")":
                while stack and stack[-1] != "(":
                    result.append(stack.pop())
                stack.pop()
            else:
                while stack and self.precedence(stack[-1]) >= self.precedence(char):
                    result.append(stack.pop())
                stack.append(char)
        
        while stack:
            result.append(stack.pop())
        
        return "".join(result)

s = "a*(b+c)/d"
print(Solution().infixtoPostfix(s))
````

# Infix to Prefix Conversion

## Problem Statement
Convert an infix expression to its prefix form.

## Description
You are given a string `s` representing an infix expression. Convert this infix expression to a prefix expression.

- **Infix expression**: The expression of the form `a op b`. When an operator is in between every pair of operands.
- **Prefix expression**: The expression of the form `op a b`. When an operator comes before its two operands.

**Note**: The precedence order is as follows: `^` has the highest precedence and is evaluated from right to left. `*` and `/` come next with left to right associativity, and `+` and `-` have the lowest precedence with left to right associativity.

## Examples
**Input:** `s = "a*(b+c)/d`  
**Output:** `/*a+bcd`  
**Explanation:** The infix expression is `a*(b+c)/d`. First, inside the brackets, `b + c` becomes `+bc`. Now the expression looks like `a*(+bc)/d`. Next, multiply `a` with `(+bc)`, so it becomes `*a+bc`. Finally, divide this result by `d`, so it becomes `/*a+bcd`.

**Input:** `s = "(a-b/c)*(a/k-l)"`  
**Output:** `*-a/bc-/akl`  
**Explanation:** The infix expression is `(a-b/c)*(a/k-l)`. First, inside the brackets, `b/c` becomes `/bc` and `a/k` becomes `/ak`. Now the expression looks like `(a-/bc) * (/ak-l)`. Next, handle the subtractions: `a-/bc` becomes `-a/bc`, and `/ak-l` becomes `-/akl`. Finally, multiply the two results: `(-a/bc * -/akl)` becomes `*-a/bc-/akl`.

## Constraints
- 3 ≤ `s.length()` ≤ 5*10³
- `s[i]` can be an operand (a–z, A–Z, 0–9), an operator (+, -, *, /, ^) or a parenthesis ((, ))

### Intuition and Approach
Prefix places operators before operands. To convert from infix, reverse the infix string (adjusting parentheses), convert to postfix, then reverse the result. This leverages the postfix conversion logic while handling right-to-left evaluation.

### Code Explanation
- Reverse `s`, swap '(' and ')' to maintain grouping.
- Use stack for operators: process left to right on reversed string.
- For operands, append to result; for operators, pop higher precedence (adjusted for right assoc of ^).
- Reverse final result for prefix.

### Dry Run
Input: "(a-b/c)*(a/k-l)"
- Reverse: )l-k/a(* )c/b-a(
- Swap parens: (l-k/a)*(c/b-a)
- Process as infix to postfix on this.
- Result: l k - a / * c b / a - *
- Reverse: *-a/bc-/akl

### Potential Edge Cases
- Simple: "a+b" → "+ab"
- Parentheses: "(a+b)" → "+ab"
- Precedence: "a*b+c" → "+*abc"
- Invalid: Assumes valid.

### Time and Space Complexity
- Time: O(N) - reversals and scan.
- Space: O(N).
- All cases O(N).

````python
# Infix to Prefix
class Solution:
    def precedence(self,ch):
        if ch == "+" or ch == "-":
            return 1
        if ch == "*" or ch == "/":
            return 2
        if ch == "^":
            return 3
        return 0

    def infixToPrefix(self, s):
        revS = ""
        for i in range(len(s)-1,-1,-1):
            if s[i] == "(":
                revS += ")"
            elif s[i] == ")":
                revS += "("
            else:
                revS += s[i]
        
        result = []
        stack = []

        for char in revS:
            if ("a" <= char <= "z") or ("A" <= char <= "Z") or ("0" <= char <= "9"):
                result.append(char)
            elif char == "(":
                stack.append(char)
            elif char == ")":
                while stack and stack[-1] != "(":
                    result.append(stack.pop())
                stack.pop()
            else:
                while stack and self.precedence(stack[-1]) > self.precedence(char):
                    result.append(stack.pop())
                stack.append(char)
            
        while stack:
            result.append(stack.pop())
        
        return "".join(result[::-1])

s = "(a-b/c)*(a/k-l)"
print(Solution().infixToPrefix(s))
````

# Prefix to Postfix Conversion

## Problem Statement
Convert a prefix expression to its postfix form.

## Description
You are given a string that represents the prefix form of a valid mathematical expression. Convert it to its postfix form.

## Example
**Input:** *-A/BC-/AKL  
**Output:** ABC/-AK/L-*  
**Explanation:** The above output is its valid postfix form.

## Task
Complete the function `preToPost(string pre_exp)`, which takes a prefix string as input and returns its postfix form.

## Complexity
- **Expected Time Complexity:** O(N)
- **Expected Auxiliary Space:** O(N)

## Constraints
- 3 ≤ pre_exp.length() ≤ 100

### Intuition and Approach
Prefix notation places operators before operands. To convert to postfix, process from right to left: push operands, on operator, pop two operands, combine as operand1 + operand2 + operator.

### Code Explanation
- stack = []
- for i in range(len(pre_exp)-1, -1, -1):
  - if pre_exp[i].isalnum(): stack.append(pre_exp[i])
  - else: operand1 = stack.pop(), operand2 = stack.pop(), new_expr = operand1 + operand2 + pre_exp[i], stack.append(new_expr)
- return stack[-1]

### Dry Run with All Steps
Input: "*-A/BC-/AKL"
- i=10 'L': append 'L' → ['L']
- i=9 'K': append 'K' → ['L','K']
- i=8 'A': append 'A' → ['L','K','A']
- i=7 '/': operand1='A', operand2='K', 'AK/', append → ['L','AK/']
- i=6 '-': operand1='AK/', operand2='L', 'AK/L-', append → ['AK/L-']
- i=5 'C': append 'C' → ['AK/L-','C']
- i=4 'B': append 'B' → ['AK/L-','C','B']
- i=3 '/': operand1='B', operand2='C', 'BC/', append → ['AK/L-','BC/']
- i=2 'A': append 'A' → ['AK/L-','BC/','A']
- i=1 '-': operand1='A', operand2='BC/', 'ABC/-', append → ['AK/L-','ABC/-']
- i=0 '*': operand1='ABC/-', operand2='AK/L-', 'ABC/-AK/L-*', append → ['ABC/-AK/L-*']
- return 'ABC/-AK/L-*'

### Potential Edge Cases
- Empty: Error
- Single operand: Returns operand
- No operators: Last operand
- Long: Handles up to 100

### Time and Space Complexity
- Time: O(N)
- Space: O(N)
- Best/Worst/Average: O(N)

````python
class Solution:
    def preToPost(self, pre_exp):
        stack = []

        for i in range(len(pre_exp)-1,-1,-1):
            if pre_exp[i].isalnum():
                stack.append(pre_exp[i])
            else:
                operand1 = stack.pop()
                operand2 = stack.pop()
                new_expr = operand1 + operand2 + pre_exp[i]
                stack.append(new_expr)
        return stack[-1]


s = "*-A/BC-/AKL"
print(Solution().preToPost(s))
````

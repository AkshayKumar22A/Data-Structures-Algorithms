### 🧠 Intuition and Approach

**Valid Parentheses** is a classic problem that requires checking if brackets are properly matched and nested. The key insight is that **parentheses must be closed in the reverse order they were opened**, which perfectly matches the **LIFO (Last In First Out)** property of a stack.

**Core Idea:**
- Use a stack to track opening brackets as we encounter them
- When we see a closing bracket, it should match the most recent opening bracket (top of stack)
- If it matches, pop the opening bracket; if not, or if stack is empty, the string is invalid
- At the end, the stack should be empty (all brackets matched)

**Why Stack?**
- Stack naturally handles the "last opened, first closed" requirement
- O(1) push/pop operations make it efficient
- We can use Python's `collections.deque` for optimal performance

### 💡 Code Explanation

```python
from collections import deque
class Solution:
    def isValid(self, s: str) -> bool:
        stack = deque()
```
- Import `deque` for efficient stack operations (O(1) append/pop)
- Create empty stack to track opening brackets
- Method signature follows LeetCode format

```python
        for i in range(len(s)):
```
- Iterate through each character in the string `s`
- Use index `i` to access `s[i]` (though we could iterate directly over characters)

```python
            if s[i] == "[" or s[i] == "(" or s[i] == "{":
                stack.append(s[i])
```
- **Opening brackets**: Push them onto the stack
- These are the brackets we expect to close later
- `append()` adds to the right end (top of stack)

```python
            else:
```
- **Closing brackets**: Handle `)`, `]`, `}`
- The `else` catches all non-opening characters (which should be closing brackets)

```python
                if len(stack) == 0:
                    return False
```
- **Critical check**: If stack is empty but we have a closing bracket, it's invalid
- No matching opening bracket exists
- Return `False` immediately (no point continuing)

```python
                ch = stack.pop()
```
- Pop the most recent opening bracket from stack
- This should be the matching opening bracket for the current closing one

```python
                if ((s[i]== ")" and ch == "(")
                    or (s[i] == "]" and ch == "[")
                    or (s[i] == "}" and ch == "{")):
                    continue
                else:
                    return False
```
- **Match check**: Verify the popped opening bracket matches the current closing bracket
- If match: `continue` to next character (valid so far)
- If no match: `return False` (mismatched brackets)
- The nested `if` conditions check all three bracket types

```python
        return len(stack) == 0
```
- **Final check**: After processing all characters, stack should be empty
- If stack has leftover opening brackets, they're unmatched → invalid
- If empty, all brackets were properly matched → valid

### ⚠️ Potential Edge Cases

#### Case 1: Empty String
```python
s = ""
# Loop doesn't run, stack = []
# return len(stack) == 0 → True
```
**Valid:** Empty string has no brackets, so it's valid.

#### Case 2: Single Opening Bracket
```python
s = "("
# Push "(" → stack = ["("]
# End: len(stack) == 0? False → return False
```
**Invalid:** Unmatched opening bracket.

#### Case 3: Single Closing Bracket
```python
s = ")"
# Closing bracket, len(stack) == 0 → return False
```
**Invalid:** No matching opening bracket.

#### Case 4: Nested Brackets
```python
s = "([{}])"
# Push (, [, {, } → pop {, [ → pop [, ( → pop ( → stack = []
# return True
```
**Valid:** Proper nesting.

#### Case 5: Wrong Order
```python
s = "([)]"
# As shown in dry run: fails at ")" expecting "(" but gets "["
```
**Invalid:** Wrong closing order.

#### Case 6: Mixed Types
```python
s = "([)]"
# Fails as above
```
**Invalid:** Type mismatch.

#### Case 7: Maximum Length (10^4 characters)
- Algorithm handles up to 10^4 characters efficiently
- No special handling needed; stack grows to O(n) in worst case

#### Case 8: All Opening Brackets
```python
s = "(([[{{"
# All pushed to stack, len(stack) > 0 → False
```
**Invalid:** All unmatched.

#### Case 9: All Closing Brackets
```python
s = "))]}"
# First ")" fails immediately (empty stack)
```
**Invalid:** No openings.

#### Case 10: Zero Input (empty string)
- Handled as valid (see Case 1)

**Note:** No "negative numbers" since input is string of brackets only.

### ⏱️ Time & Space Complexity

| Operation/Case | Best Case | Average Case | Worst Case | Explanation |
|----------------|-----------|--------------|-----------|-------------|
| **Time Complexity** | O(n) | O(n) | O(n) | Single pass through string; each character processed once with O(1) stack operations |
| **Space Complexity** | O(1) | O(n) | O(n) | Stack uses O(n) space in worst case (all opening brackets); O(1) if no openings |

**Detailed Analysis:**
- **Time:** O(n) where n = len(s). Each character is visited exactly once, and stack operations are O(1).
- **Space:** O(n) worst case when all characters are opening brackets (e.g., "(([[{{"); O(1) best case for empty or balanced strings with no nesting.
- **Average:** O(n) space for typical nested structures.

**Why O(1) best case space?** For strings like "()" or "[]{}", stack empties frequently, but amortized it's still O(n).

### 📝 Key Insights

- **Stack is essential:** Without it, tracking nested brackets would be impossible
- **Early termination:** Checking empty stack for closing brackets prevents unnecessary processing
- **Match validation:** The three-way condition ensures type-specific matching
- **Final stack check:** Catches unmatched opening brackets
- **Efficiency:** Linear time makes it suitable for large inputs (up to 10^4)

This solution is optimal and follows the standard approach for parentheses validation problems. The use of `deque` provides the best performance for stack operations in Python.

### 🔗 Quick Reference

```
Valid Cases:   ()  []  {}  ()[]{}  ([{}])  {([])}
Invalid Cases: (]  ([)]  ((  ))  ]  }{  ([}
```

```
Algorithm Steps:
1. Initialize empty stack
2. For each character:
   - If opening: push to stack
   - If closing: check stack not empty, pop, verify match
3. Return stack.is_empty()
```

````python
from collections import deque
class Solution:
    def isValid(self, s: str) -> bool:
        stack = deque()
        for i in range(len(s)):
            if s[i] == "[" or s[i] == "(" or s[i] == "{":
                stack.append(s[i])
            else:
                if len(stack) == 0:
                    return False
                ch = stack.pop()
                if ((s[i]== ")" and ch == "(")
                    or (s[i] == "]" and ch == "[")
                    or (s[i] == "}" and ch == "{")):
                    continue
                else:
                    return False
        return len(stack) == 0

s = "()"
print(Solution().isValid(s))
s = "()[]{}"
print(Solution().isValid(s))  
s = "(]"
print(Solution().isValid(s))  
s = "([])"
print(Solution().isValid(s))  
s = "([)]"
print(Solution().isValid(s))  
s = "]"
print(Solution().isValid(s))
s = "){"
print(Solution().isValid(s))  
````

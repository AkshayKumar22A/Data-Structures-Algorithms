## Problem Statement

You are given an integer array `plants` where `plants[i]` is the amount of water the i-th plant needs. There is a water can with capacity `capacity`. A gardener waters the plants in order from left to right starting at the river (position -1). When the can does not have enough water for the next plant, the gardener returns to the river to refill (walking back and forth costs steps). The task is to compute the total number of steps the gardener takes to water all plants.

---

## Intuition and Approach
Walk through the plants from left to right, maintaining the remaining water in the can. For each plant:
- If remaining water >= water needed by the plant, water it and subtract the amount.
- Otherwise, go back to river to refill (which costs walking distance to river and back), then water the plant.

Count steps as follows:
- Moving from plant i-1 to plant i costs 1 step.
- When refilling, the additional cost is `2 * i` (go from plant i to river and back to plant i).

This greedy simulation runs in a single pass.

---

## Code Explanation in Details
Assume the code (in the next code cell) follows this pattern:

1. Initialize `steps = 0` and `water = capacity`.
2. For each index `i` from 0 to len(plants)-1:
    - Increment `steps` by 1 for moving to the plant (from previous position; the first plant also costs 1 step from the river).
    - If `water < plants[i]`, then add `2 * i` to `steps` (go back to river and return) and reset `water = capacity`.
    - Water the plant: `water -= plants[i]`.
3. Return `steps`.

Note: Some implementations add movement steps differently (e.g., adding 1 per plant traversal then adding refill cost). The important invariant is accounting for both forward movement and refill detours.

---

## Dry Run (Edge Case Example)
Input: `plants = [2,2,3,3]`, `capacity = 5`

Start: steps=0, water=5
- i=0 (plant 0 needs 2): move to plant → steps=1; water(5) >=2 → water=3
- i=1 (needs 2): move → steps=2; water(3) >=2 → water=1
- i=2 (needs 3): move → steps=3; water(1) <3 → refill: steps += 2*2 =4 → steps=7; water=5; water-=3 → water=2
- i=3 (needs 3): move → steps=8; water(2) <3 → refill: steps += 2*3 =6 → steps=14; water=5; water-=3 → water=2
End: total steps = 14

This matches the standard simulation answers.

---

## Edge Cases
- **Zero input:** `plants = []` → steps = 0.
- **All plants require 0 water:** `plants = [0,0,0]` → steps = number of plants (only forward moves), no refills.
- **Single plant larger than capacity:** If any `plants[i] > capacity` the problem constraints usually prevent this; otherwise it's impossible to water that plant (or you must allow multiple refills per plant). Assume constraints ensure `plants[i] <= capacity`.
- **Capacity exactly equals total needed:** No refill needed except initial fill.

---

## Time and Space Complexity
- **Time Complexity:** O(n) — single pass over `plants`.
- **Space Complexity:** O(1) — constant extra space for counters.

Exact counts:
- For each plant we perform O(1) work (a few arithmetic ops and comparisons). So total steps are proportional to `n` operations.

```python
from typing import List
class Solution:
    def wateringPlants(self, plants: List[int], capacity: int) -> int:
        l = len(plants)
        water = capacity
        ans = 0
        start = 0
        while start<l :
            if water >= plants[start]:
                water -= plants[start]
                ans +=1
                start+=1
            else:
                water = capacity
                ans+=start*2
        return ans

plants = [2,2,3,3]
capacity = 5
s = Solution()
s.wateringPlants(plants,capacity)
```

## Problem Statement

Given two sorted integer arrays `nums1` and `nums2`, and their respective counts `m` and `n`, merge `nums2` into `nums1` as one sorted array in-place. `nums1` has enough space (size `m + n`) to hold additional elements from `nums2`.

---

## Intuition and Approach

Since both arrays are sorted, we can merge them efficiently from the end to avoid overwriting elements in `nums1`. We use three pointers:
- `l1` for the last valid element in `nums1`
- `l2` for the last element in `nums2`
- `i` for the last position in `nums1`

Compare elements from the end and place the larger one at position `i`, moving pointers accordingly.

---

## Code Explanation in Details

1. Initialize `l1 = m - 1`, `l2 = n - 1`, and `i = m + n - 1`.
2. While `l2 >= 0`:
    - If `l1 >= 0` and `nums1[l1] > nums2[l2]`, set `nums1[i] = nums1[l1]`, decrement `l1` and `i`.
    - Else, set `nums1[i] = nums2[l2]`, decrement `l2` and `i`.
3. Continue until all elements from `nums2` are merged.

This ensures all elements are merged in sorted order, in-place.

---

## Dry Run (Edge Case Example)

Input: `nums1 = [1,2,3,0,0,0]`, `m = 3`, `nums2 = [2,5,6]`, `n = 3`

- l1 = 2, l2 = 2, i = 5
- nums1[2]=3 > nums2[2]=6 → False
- nums1[5]=6 (from nums2), l2=1, i=4
- nums1[2]=3 > nums2[1]=5 → False
- nums1[4]=5 (from nums2), l2=0, i=3
- nums1[2]=3 > nums2[0]=2 → True
- nums1[3]=3 (from nums1), l1=1, i=2
- nums1[1]=2 > nums2[0]=2 → False
- nums1[2]=2 (from nums2), l2=-1, i=1

Result: `[1,2,2,3,5,6]`

---

## Edge Cases

- **Zero input:** `nums1 = [0], m = 0, nums2 = [1], n = 1` → Output: `[1]`
- **All elements in nums2 smaller:** `nums1 = [4,5,6,0,0,0], m = 3, nums2 = [1,2,3], n = 3` → Output: `[1,2,3,4,5,6]`
- **All elements in nums2 larger:** `nums1 = [1,2,3,0,0,0], m = 3, nums2 = [4,5,6], n = 3` → Output: `[1,2,3,4,5,6]`
- **Negative numbers:** `nums1 = [-3,-2,-1,0,0,0], m = 3, nums2 = [-5,-4,-2], n = 3` → Output: `[-5,-4,-3,-2,-2,-1]`
- **Single element:** `nums1 = [0], m = 0, nums2 = [1], n = 1` or `nums1 = [1], m = 1, nums2 = [], n = 0`

---

## Time and Space Complexity

- **Time Complexity:** $O(m + n)$
  - Each element is visited once, so total operations are $m + n$.
- **Space Complexity:** $O(1)$
  - In-place merge, no extra space used.

**Calculation:**
- For arrays of length $m$ and $n$, each element is compared and placed once.
- $O(m + n)$ steps, no additional space required.


```python
from typing import List
class Solution:
    def merge(self, nums1: List[int], m: int, nums2: List[int], n: int) -> None:
        """
        Do not return anything, modify nums1 in-place instead.
        """
        l1 = m - 1
        l2 = n - 1
        i = m+n - 1
        while(l2 >= 0):
            if l1 >= 0 and nums1[l1] > nums2[l2]:
                nums1[i] = nums1[l1]
                i-=1
                l1-=1
            else:
                nums1[i] = nums2[l2]
                i-=1
                l2-=1
        return nums1

nums1 = [1,2,3,0,0,0]
m = 3
nums2 = [2,5,6]
n = 3

s = Solution()
s.merge(nums1,m,nums2,n)
```

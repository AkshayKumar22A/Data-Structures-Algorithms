Check if Array Is Sorted and Rotated</a></h2><h3>Easy</h3><hr><p>Given an array <code>nums</code>, return <code>true</code><em> if the array was originally sorted in non-decreasing order, then rotated <strong>some</strong> number of positions (including zero)</em>. <p>There may be <strong>duplicates</strong> in the original array.</p>

<p><strong>Note:</strong> An array <code>A</code> rotated by <code>x</code> positions results in an array <code>B</code> of the same length such that <code>B[i] == A[(i+x) % A.length]</code> for every valid index <code>i</code>.</p>

<strong>Output:</strong> true

</pre>

<p><strong class="example">Example 2:</strong></p>

<pre>
<strong>Input:</strong> nums = [2,1,3,4]
<strong>Output:</strong> false
<strong>Explanation:</strong> There is no sorted array once rotated that can make nums.
</pre>

<p><strong class="example">Example 3:</strong></p>

<pre>
<strong>Input:</strong> nums = [1,2,3]
<strong>Output:</strong> true

You can rotate the array by x = 0 positions (i.e. no rotation) to make nums.
</pre>

## Intuition and Approach
The problem asks if an array is a rotation of a non-decreasing sorted array. That means, after some number of rotations, the array should look sorted except for a single place where the order "wraps around".

**Key Insight:**
- In a sorted and rotated array, there can be at most one place where `nums[i] > nums[(i+1) % n]` (the rotation point).
- If there are more than one such places, the array cannot be a rotation of a sorted array.

---

## Code Explanation in Detail
- Loop through the array and count the number of places where the next element is smaller than the current (`nums[i] > nums[(i+1)%n]`).
- If this count is more than 1, return `False`.
- Otherwise, return `True`.
- The modulo operation allows the check to "wrap around" from the last element to the first.

---

## Dry Run (Edge Case)
Test case: `nums = [3,4,5,1,2]`
- Compare 3 vs 4: OK
- Compare 4 vs 5: OK
- Compare 5 vs 1: 5 > 1 (rotation point, count=1)
- Compare 1 vs 2: OK
- Compare 2 vs 3: 2 < 3 (no rotation point)
- Only one rotation point → Output: **True**

Test case: `nums = [2,1,3,4]`
- Compare 2 vs 1: 2 > 1 (rotation point, count=1)
- Compare 1 vs 3: OK
- Compare 3 vs 4: OK
- Compare 4 vs 2: 4 > 2 (second rotation point, count=2)
- More than one rotation point → Output: **False**

Test case: `nums = [1,2,3]`
- Compare 1 vs 2: OK
- Compare 2 vs 3: OK
- Compare 3 vs 1: 3 > 1 (rotation point, count=1)
- Only one rotation point (or zero if not rotated) → Output: **True**

---

## Edge Cases
- **Zero input:** Not allowed by constraints (minimum length is 1)
- **Single element:** Always True (trivially sorted and rotated)
- **All elements same:** Always True (no rotation point)
- **Negative numbers:** Not possible by constraints (all elements ≥ 1)
- **Duplicates:** Allowed; only count strict decreases

---

## Time and Space Complexity
- **Time Complexity:**
    - Best: $O(n)$ (single pass)
    - Average: $O(n)$
    - Worst: $O(n)$
    - Exact: $O(n)$
- **Space Complexity:**
    - Best/Average/Worst: $O(1)$ (constant extra space)
    - Exact: $O(1)$

---

## Summary
- The solution checks for at most one rotation point.
- Handles all edge cases efficiently in $O(n)$ time and $O(1)$ space.


```python
from typing import List
class Solution:
    def check(self, nums: List[int]) -> bool:
        l = len(nums)
        rotations = 0
        for i in range(l):
            if nums[i] > nums[(i+1)%l]:
                rotations +=1
            if rotations > 1:
                return False
        return True

nums = [3,4,5,1,2]
s = Solution()
s.check(nums)
```

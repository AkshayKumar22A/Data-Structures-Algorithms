## Intuition and Approach
We need to count how many rods (0-9) have all three colors of rings: Red (R), Green (G), and Blue (B). For each ring, we track its color and rod. After processing all rings, we check each rod to see if it contains all three colors.

## Code Explanation in Details
- Create a tracker list of 10 strings, one for each rod.
- For each ring (pair of characters in `rings`), append its color to the corresponding rod's tracker.
- After processing, for each rod, check if its tracker contains 'R', 'G', and 'B'.
- Count rods that have all three colors.

## Dry Run (Edge Case)
Test case: `rings = "G4"`
- Only one ring: Green on rod 4.
- tracker: `['', '', '', '', 'G', '', '', '', '', '']`
- No rod has all three colors.
- Output: **0**

Test case: `rings = "B0B6G0R6R0R6G9"`
- Process each pair:
    - B0 → tracker[0] += 'B'
    - B6 → tracker[6] += 'B'
    - G0 → tracker[0] += 'G'
    - R6 → tracker[6] += 'R'
    - R0 → tracker[0] += 'R'
    - R6 → tracker[6] += 'R'
    - G9 → tracker[9] += 'G'
- tracker: `['BGR', '', '', '', '', '', 'BRR', '', '', 'G']`
- Only rod 0 has all three colors.
- Output: **1**

## Edge Cases
- **Zero input:** `rings = ""` → Output: 0
- **Single ring:** No rod can have all three colors
- **Multiple rings, but not all colors:** Output: 0
- **Negative numbers:** Not possible (rods are 0-9)
- **Duplicate colors on same rod:** Still only counts if all three colors are present

## Time and Space Complexity
- **Time Complexity:**
    - Best: $O(n)$ (where $n$ is length of `rings`)
    - Average: $O(n)$
    - Worst: $O(n)$
    - Exact: $O(n)$
- **Space Complexity:**
    - Best: $O(1)$ (tracker size is fixed at 10)
    - Average: $O(1)$
    - Worst: $O(1)$
    - Exact: $O(1)$

## Intuition & Approach
An **anagram** means that two strings contain the exact same characters with the same frequency, but possibly in a different order.
Given the constraints that the input strings only contain lowercase English letters, we can leverage frequency counting to compare the two words efficiently.

### Step-by-step strategy
1. **Length check** – if the lengths differ, the strings cannot be anagrams.
2. **Build frequency maps** for both strings. Iterate over each character and count how many times it appears.
3. **Compare the maps** – ensure every character in the first map appears with identical frequency in the second.
4. If all counts match, the strings are anagrams; otherwise return `False`.

## Code Explanation
Below is the Python implementation that follows the above approach. Each line is annotated with the reasoning behind it.
```python
class Solution:
    def isAnagram(self, s: str, t: str) -> bool:
        # 1. Quick reject: lengths must match
        if len(s) != len(t):
            return False
        
        # 2. Build frequency map for s
        freq_maps = dict()  # keys are chars, values are counts
        for c in s:  # iterate over each character in s
            # increment the count for this character; default 0 if unseen
            freq_maps[c] = freq_maps.get(c,0) + 1
        
        # 3. Build frequency map for t in the same way
        freq_mapt = dict()
        for c in t:
            freq_mapt[c] = freq_mapt.get(c,0) + 1

        # 4. Compare the two maps character by character
        for i in freq_maps:  # only need to iterate keys of the first map
            if i not in freq_mapt:  # missing character in t
                return False
            elif freq_maps[i] != freq_mapt[i]:  # different count
                return False

        # if we reach here, frequencies match for every character
        return True
```

### Why each step?
- The length check saves time: two strings of different sizes cannot be anagrams.
- Frequency maps give us `O(n)` comparisons rather than sorting which costs `O(n log n)`.
- Comparing maps avoids constructing a third data structure; we simply walk one dictionary and probe the other.

## Dry Run
Take `s = "anagram"`, `t = "nagaram"`:
1. lengths both 7 ⇒ continue.
2. Build `freq_maps`: `{a:3, n:1, g:1, r:1, m:1}`.
3. Build `freq_mapt`: the same distribution.
4. Iterate over `freq_maps` keys: each character exists in `freq_mapt` with equal count.
5. Return True.

For a negative scenario, `s = "rat"`, `t = "car"`:
- After building maps (`{r:1,a:1,t:1}` vs `{c:1,a:1,r:1}`), the loop finds `t` missing in the second map and returns `False`.

## Edge Cases & Notes
- **Empty strings**: both `""` and `""` return `True` (lengths equal, maps empty).
- **Single character**: `"a"` vs `"a"` yields `True`; vs `"b"` gives `False`.
- **Unicode or extended charset**: the algorithm works for any hashable characters; however, Python’s `dict` handles them seamlessly. The follow‑up question suggests using `collections.Counter` or sorting when full unicode normalization is required.
- **Zero or negative numbers** are irrelevant here since inputs are strings; if the problem were numbers, ensure type validation.

## Time & Space Complexity
| Case | Time | Space |
|------|------|-------|
| Best | `O(1)` length mismatch early exit (short strings) | `O(1)` additional space |
| Average/Worst | `O(n)` to scan both strings and compare maps | `O(n)` for frequency dictionaries |

- **Sorting alternative**: would cost `O(n log n)` time but `O(n)` space as well. The chosen map-based approach is optimal for this problem given the constraints.

---


````python
class Solution:
    def isAnagram(self, s: str, t: str) -> bool:
        if len(s) != len(t):
            return False
        
        freq_maps = dict()
        for c in s:
            freq_maps[c] = freq_maps.get(c,0) + 1
        
        freq_mapt = dict()
        for c in t:
            freq_mapt[c] = freq_mapt.get(c,0) + 1

        for i in freq_maps:
            if i not in freq_mapt:
                return False
            elif freq_maps[i] != freq_mapt[i]:
                return False

        return True
````

## Single-pass frequency counts

### Intuition & Approach
The goal remains to compare character frequencies. Instead of building one map at a time, we can walk both strings in parallel and update two dictionaries simultaneously. This does **not** change complexity, but it reduces the number of loops and can feel more elegant in code.

1. **Length check** – different lengths imply `False`.  
2. Create two empty maps, `countS` and `countT`.  
3. Iterate over indices `i` from `0` to `len(s)-1`; at each step increment counters for `s[i]` and `t[i]`.  
4. After the loop, compare the two maps directly; if equal, return `True`, otherwise `False`.

### Code Explanation
```python
class Solution:
    def isAnagram(self, s: str, t: str) -> bool:
        # 1. ensure same length
        if len(s) != len(t):
            return False

        # 2. initialize frequency dictionaries
        countS, countT = {}, {}

        # 3. simultaneously count characters from both strings
        for i in range(len(s)):
            # increment count for s[i]
            countS[s[i]] = 1 + countS.get(s[i], 0)
            # increment count for t[i]
            countT[t[i]] = 1 + countT.get(t[i], 0)

        # 4. dictionary equality checks both keys and values
        return countS == countT
```

- Both `get` calls default to `0` when the key isn’t present; this handles first-time characters gracefully.
- The final comparison uses Python’s built-in dictionary equality, which is efficient and clear.

### Dry Run
For `s = "anagram"`, `t = "nagaram"`:  
- After length check, both maps get built in parallel: `{a:3,n:1,g:1,r:1,m:1}` for both.  
- Final equality returns `True`.

For `s = "rat"`, `t = "car"`:  
- After iteration, `countS={r:1,a:1,t:1}`, `countT={c:1,a:1,r:1}`.  
- Comparison fails (`t` missing, `c` extra) and returns `False`.

### Edge Cases & Notes
- Works with empty strings (`{}` == `{}` → `True`).
- Handles single-character strings identically to the first method.
- No new edge cases beyond those already discussed.

### Time & Space Complexity
| Case                 | Time       | Space        |
|---------------------|------------|--------------|
| Best (length mismatch) | `O(1)`     | `O(1)`       |
| Average/Worst       | `O(n)`     | `O(n)`       |

Complexities are the same as the prior solution; the difference is stylistic rather than algorithmic.

---

````python
class Solution:
    def isAnagram(self, s: str, t: str) -> bool:
        if len(s) != len(t):
            return False

        countS, countT = {}, {}

        for i in range(len(s)):
            countS[s[i]] = 1 + countS.get(s[i], 0)
            countT[t[i]] = 1 + countT.get(t[i], 0)
        return countS == countT
````

## Sorting-based one-liner

### Intuition & Approach
If two strings are anagrams, they contain the same characters in a different order. Sorting each string arranges the characters into a canonical order; identical sorted sequences indicate an anagram. This avoids explicit counting at the cost of the sort operation.

Procedure:
1. Check length; if different, return `False`.  
2. Sort both strings using `sorted()`.  
3. Compare the resulting lists; return the boolean result directly.

### Code Explanation
```python
class Solution:
    def isAnagram(self, s: str, t: str) -> bool:
        # sort both inputs and compare
        return True if sorted(s) == sorted(t) else False
```
- `sorted(s)` converts `s` to a list of characters and sorts them.
- The comparison operator `==` on lists returns `True` only if they are the same length and all corresponding elements match.
- The ternary is optional; one could simply write `return sorted(s) == sorted(t)`.

### Dry Run
With `s = "anagram"` and `t = "nagaram"`:  
- Sorted results: `['a','a','a','g','m','n','r']` for both.  
- Lists equal → return `True`.

With `s = "rat"` and `t = "car"`:  
- Sorted `s` → `['a','r','t']`; sorted `t` → `['a','c','r']`.  
- Lists differ (`'t'` vs `'c'`) → return `False`.

### Edge Cases & Notes
- Empty strings sort to empty lists; equality holds.  
- Works for any comparable characters (including Unicode).  
- Sorting implicitly handles frequency by arrangement, so no separate map is needed.

### Time & Space Complexity
| Case              | Time           | Space        |
|------------------|----------------|--------------|
| All              | `O(n log n)`   | `O(n)`       |

While simpler, this method is not optimal for very long strings; mention this during interviews if performance is a concern.

---
Sorting is a great fallback or a one-liner for quick solutions but should be justified against the linear approaches when asked about efficiency.

````python
class Solution:
    def isAnagram(self, s: str, t: str) -> bool:
        return True if sorted(s) == sorted(t) else False
````

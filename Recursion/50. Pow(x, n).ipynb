```python
class Solution:
    def myPow(self, x: float, n: int) -> float:
        self.ans = 1.0
        self.solve(self.ans,x,n)
        return self.ans 
    
    def solve(self,ans,x,n):
        if n == 0:
            return
        self.ans = self.ans * x
        n -= 1
        self.solve(ans,x,n)


x = 2.00000 
n = 2
sol = Solution().myPow(x,n)
print(sol)
```

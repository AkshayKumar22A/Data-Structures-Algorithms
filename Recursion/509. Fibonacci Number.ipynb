# Fibonacci Number: Recursive Solution

---

## 1. Problem Statement
Given an integer $n$, compute the $n$th Fibonacci number, where:
- $F(0) = 0$
- $F(1) = 1$
- $F(n) = F(n-1) + F(n-2)$ for $n > 1$

## 2. Intuition and Approach
The Fibonacci sequence is naturally defined recursively. For each $n$, the value is the sum of the previous two Fibonacci numbers. Recursion is a direct way to implement this definition.

## 3. Code Explanation in Details
```python
class Solution:
    def fibonacci(self, n: int) -> int:
        if n == 0:
            return 0
        elif n == 1:
            return 1
        else:
            return self.fibonacci(n-1) + self.fibonacci(n-2)

n = 10
fib = Solution()
fib.fibonacci(n)
```
- If $n$ is $0$ or $1$, return $n$ (base cases).
- Otherwise, recursively compute $F(n-1)$ and $F(n-2)$ and sum them.

## 4. Dry Run
For $n = 4$:
- fibonacci(4) = fibonacci(3) + fibonacci(2)
- fibonacci(3) = fibonacci(2) + fibonacci(1)
- fibonacci(2) = fibonacci(1) + fibonacci(0)
- fibonacci(1) = 1
- fibonacci(0) = 0
- Unwind: fibonacci(2) = 1 + 0 = 1
- fibonacci(3) = 1 + 1 = 2
- fibonacci(4) = 2 + 1 = 3

## 5. Edge Cases
- $n = 0$: Output is $0$
- $n = 1$: Output is $1$
- Negative $n$: Not handled; should add input validation.
- Large $n$: Recursion is inefficient and may cause stack overflow.

## 6. Time and Space Complexity
- **Time Complexity:** $O(2^n)$ (exponential, due to repeated calculations)
- **Space Complexity:** $O(n)$ (recursion stack depth)

---

````python
class Solution:
    def fibonacci(self,n: int)-> int:
        if n == 0:
            return 0
        elif n == 1:
            return 1
        else:
            return self.fibonacci(n-1) + self.fibonacci(n-2)

n = 10
fib = Solution()
fib.fibonacci(n)
````


# Fibonacci Number: Iterative Solution

---

## 1. Problem Statement
Given an integer $n$, print the first $n$ Fibonacci numbers.

## 2. Intuition and Approach
Instead of recursion, use a loop to iteratively compute each Fibonacci number, updating two variables to hold the previous values.

## 3. Code Explanation in Details
```python
n = 11
a = 0
b = 1
c = a + b
for i in range(1, n):
    print(c)
    c = a + b
    a = b
    b = c
```
- Initialize $a = 0$, $b = 1$.
- For each iteration, compute $c = a + b$, print $c$, then update $a$ and $b$.
- Prints the sequence up to the $n$th number.

## 4. Dry Run
For $n = 5$:
- Iteration 1: $a=0$, $b=1$, $c=1$ → print 1
- Iteration 2: $a=1$, $b=1$, $c=2$ → print 2
- Iteration 3: $a=1$, $b=2$, $c=3$ → print 3
- Iteration 4: $a=2$, $b=3$, $c=5$ → print 5

## 5. Edge Cases
- $n = 0$: No output
- $n = 1$: Only the first Fibonacci number
- Negative $n$: Not handled; should add input validation.

## 6. Time and Space Complexity
- **Time Complexity:** $O(n)$
- **Space Complexity:** $O(1)$ (constant space for variables)

````python
n = 10
a = 0
b = 1
c = a + b
for i in range(1,n):
    #print(c) # Uncomment to print the series
    c = a + b
    a = b
    b = c
print(c)
````

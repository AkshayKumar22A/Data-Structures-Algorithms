<h2><a href="https://leetcode.com/problems/combination-sum-iii/">216. Combination Sum III</a></h2><h3>Medium</h3><hr><p>Find all valid combinations of <code>k</code> numbers that sum up to <code>n</code> such that the following conditions are true:</p>

<ul>
	<li>Only numbers <code>1</code> through <code>9</code> are used.</li>
	<li>Each number is used <strong>at most once</strong>.</li>
</ul>

<p>Return <em>a list of all possible valid combinations</em>. The list must not contain the same combination twice, and the combinations may be returned in any order.</p>

<pre>
<strong>Input:</strong> k = 3, n = 7
<strong>Output:</strong> [[1,2,4]]
<strong>Explanation:</strong>
1 + 2 + 4 = 7
There are no other valid combinations.</pre>

<p><strong class="example">Example 2:</strong></p>

<pre>
<strong>Input:</strong> k = 3, n = 9
<strong>Output:</strong> [[1,2,6],[1,3,5],[2,3,4]]
<strong>Explanation:</strong>
1 + 2 + 6 = 9
1 + 3 + 5 = 9
2 + 3 + 4 = 9
There are no other valid combinations.
</pre>

<p><strong class="example">Example 3:</strong></p>

<pre>
<strong>Input:</strong> k = 4, n = 1
<strong>Output:</strong> []
<strong>Explanation:</strong> There are no valid combinations.
Using 4 different numbers in the range [1,9], the smallest sum we can get is 1+2+3+4 = 10 and since 10 &gt; 1, there are no valid combination.
</pre>

</ul>


### Intuition and Approach

To solve **Combination Sum III**, we need to find all unique combinations of `k` different numbers (1–9) that sum to `n`. The natural tool for exploring combinations with constraints is **backtracking**, because it systematically builds candidate solutions and abandons (“backtracks”) when the partial solution cannot possibly lead to a valid answer.

Key ideas:
1. **Numbers 1–9** behave like a small search space, making recursion feasible.
2. **Each number used at most once** means we can maintain an index to avoid revisiting prior values and ensure combinations are generated in ascending order.
3. **Limit by `k` and `n`**: stop exploring further if either the length of the current combination exceeds `k` or the sum exceeds `n`.

### Code Explanation

```python
from typing import List
class Solution:
    def combinationSum3(self, k: int, n: int) -> List[List[int]]:
        self.result = []
        idx = 1
        self.solve(k,n,0,[],idx)
        return self.result
```
- `self.result` stores all valid combinations.
- `idx` starts at 1 because valid candidates range from 1 to 9.
- We call `self.solve()` with initial values: current sum `0`, empty path `[]`, and starting index `1`.

```python
    def solve(self,k,n,total,subarr,idx):
        if total == n and len(subarr) == k:
            self.result.append(subarr.copy())
            return
        if total > n or len(subarr) > k:
            return
```

- **Base case 1:** when the current sum equals `n` _and_ we have exactly `k` numbers, we append a copy of the path to results. Use `copy()` so future backtracking modifications don’t alter the stored combination.
- **Pruning:** if the current sum exceeds `n` or we’ve already chosen more than `k` numbers, there’s no point in continuing down this branch.

```python
        for i in range(idx,10):
            subarr.append(i)
            self.solve(k,n,total+i,subarr,i+1)
            subarr.pop()
```

- Loop from `idx` to `9` (range upper bound 10). This ensures numbers are considered in ascending order and avoids reuse.
- Add `i` to `subarr` and recurse with updated sum and next index `i+1` to avoid duplication.
- After recursion returns, `pop()` restores state for the next iteration (backtracking).

### Dry Run (k = 3, n = 7)

1. Start with `total=0`, `subarr=[]`, `idx=1`.
2. i=1 → `subarr=[1]`, call with `total=1`, `idx=2`.
   - i=2 → `subarr=[1,2]`, call with `total=3`, `idx=3`.
     * i=3 → `subarr=[1,2,3]`, sum=6, length=3 → not enough, continue.
     * i=4 → `subarr=[1,2,4]`, sum=7,length=3 → **match**; append [1,2,4].
     * i=5…9 → sum exceeds 7 or incorrect length → pruned.
   - Backtrack to [1]; i=3…9 tried similarly; none yield valid combination.
3. Backtrack to []; i=2 → [2]; similar exploration eventually gives nothing new.
4. The only valid combination found is `[1,2,4]`.

### Edge Cases

- **Zero or negative `k` or `n`**: according to problem constraints, `k≥2` and `n≥1`. Our code doesn’t explicitly handle zero/negative inputs, but it gracefully returns an empty list because no valid combination can be formed. If you need validation, add a check at the beginning of `combinationSum3`.
- **`k` too large relative to `n`**: e.g., `k=9`, `n=5`. The pruning `len(subarr) > k` or `total > n` prevents exploring useless branches.
- **`n` too large**: The maximum possible sum using 1–9 exactly once is 45; any `n > 45` will yield an empty result quickly.

### Time and Space Complexity

- **Time Complexity**: The algorithm explores combinations of up to 9 numbers. In the worst case, it generates all subsets of {1..9} of size ≤ k, giving roughly **O(2^9)** = O(512) possibilities. More precisely, it’s O(C(9, k)) per k but bounded by O(2ⁿ) with n=9. So overall we say **O(1)** with respect to input sizes since the domain is fixed, but in terms of k and n parameters it’s exponential in k.  
  - **Best case**: When pruning happens early, e.g., n small or k small, branches terminate quickly.
  - **Average case**: also bounded by a constant upper limit for n≤60, k≤9.
- **Space Complexity**: recursion depth ≤ 9 and `subarr` length ≤ 9, plus result storage. So space is **O(1)** auxiliary (or O(k) for recursion stack) and O(number of valid combinations) for the output.

````python
from typing import List
class Solution:
    def combinationSum3(self, k: int, n: int) -> List[List[int]]:
        self.result = []
        idx = 1
        self.solve(k,n,0,[],idx)
        return self.result
    
    def solve(self,k,n,total,subarr,idx):
        if total == n and len(subarr) == k:
            self.result.append(subarr.copy())
            return
        if total > n or len(subarr) > k:
            return

        for i in range(idx,10):
            subarr.append(i)
            self.solve(k,n,total+i,subarr,i+1)
            subarr.pop()

k = 3
n = 7
print(Solution().combinationSum3(k,n))
````

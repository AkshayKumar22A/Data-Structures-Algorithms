# 78. Subsets

**Difficulty:** Medium | **Topic:** Recursion & Backtracking

---

## Problem Statement

Given an integer array `nums` of **unique** elements, return **all possible subsets** (the power set).

The solution set **must not** contain duplicate subsets. Return the solution in **any order**.

### Example 1:
```
Input: nums = [1,2,3]
Output: [[],[1],[2],[1,2],[3],[1,3],[2,3],[1,2,3]]
```

### Example 2:
```
Input: nums = [0]
Output: [[],[0]]
```

### Constraints:
- `1 <= nums.length <= 10`
- `-10 <= nums[i] <= 10`
- All the numbers of `nums` are **unique**.

---

## Intuition and Approach

### Key Insight:
For every element in the array, we have **two choices**:
1. **Include** the element in the current subset
2. **Exclude** the element from the current subset

This creates a binary tree of decisions, where each path from root to leaf represents one subset.

### Why Recursion with Backtracking?
- We need to explore all possible combinations
- Backtracking allows us to "undo" choices and explore alternative paths
- Total subsets = 2^n (for n elements), which matches the binary decision tree

### Visual Representation:
For `nums = [1, 2, 3]`:
```
                           []
                    /               \
                Include 1        Exclude 1
                  /                    \
                [1]                    []
               /   \                 /   \
            Inc 2  Exc 2          Inc 2  Exc 2
            /        \            /        \
          [1,2]      [1]        [2]       []
          /   \      /  \       / \       / \
       I3 E3  I3 E3 I3 E3    I3 E3     I3 E3
       /   \  /  \  /  \    /  \      /  \
    [1,2,3][1,2][1,3][1][2,3][2][3,0][3]
```

## Code Explanation (Line-by-Line)

```python
from typing import List

class Solution:
    def subsets(self, nums: List[int]) -> List[List[int]]:
        result = []  # Line 1: Store all subsets here
        self.solve(nums, 0, [], result)  # Line 2: Start recursion
        return result  # Line 3: Return final result

    def solve(self, nums, idx, subset, result):
        # Base Case: Line 4
        if idx >= len(nums):
            # When we've processed all elements, add current subset to result
            # Why .copy()? We pass the list by reference in Python
            # Without copy(), all subsets would point to same list!
            result.append(subset.copy())
            return
        
        # Include Current Element: Line 5
        subset.append(nums[idx])  # Add current element to subset
        # Recurse with next index (include this element)
        self.solve(nums, idx + 1, subset, result)
        
        # Exclude Current Element (Backtrack): Line 6
        subset.pop()  # Remove the element we just added (backtrack)
        # Recurse with next index (exclude this element)
        self.solve(nums, idx + 1, subset, result)
```

### Parameter Explanation:
- **`nums`**: The input array
- **`idx`**: Current index we're deciding to include/exclude
- **`subset`**: Current subset being built
- **`result`**: Accumulator storing all subsets

### Why `.copy()` is Critical:
```python
# Without .copy():
result.append(subset)  # ❌ WRONG - points to same list object
# After all recursions finish, subset is empty
# So all references in result point to empty list!

# With .copy():
result.append(subset.copy())  # ✅ CORRECT - adds independent copy
# Each subset is a separate list object
```

### Step-by-Step Execution:
1. Start with empty subset `[]`
2. At each element, we create TWO branches: include or exclude
3. When `idx >= len(nums)`, we've made all decisions → save the subset
4. Backtrack by popping the last element before exploring exclude branch
5. Continue until all 2^3 = 8 subsets are generated

## Time and Space Complexity

### Time Complexity Analysis

| Case | Complexity | Explanation |
|------|-----------|-------------|
| **Best Case** | O(2^n) | Even with n=1, we visit all 2 branches |
| **Average Case** | O(2^n) | We always visit all 2^n recursive calls |
| **Worst Case** | O(2^n) | n elements = 2^n total subsets |

**Detailed Breakdown:**
- Total recursive calls = 2^n (binary tree with n levels)
- Each call does O(1) work (append/pop operations)
- Copying each subset: O(k) where k is subset size
  - Average subset size = n/2
  - But we copy 2^n times
  - Total = O(2^n × n)

**Final Time Complexity: O(n × 2^n)**
- n factor comes from copying subsets of varying lengths
- 2^n factor comes from generating all subsets

### Space Complexity Analysis

| Component | Space | Explanation |
|-----------|-------|-------------|
| **Recursion Stack** | O(n) | Maximum depth = n (for largest subset) |
| **Output Storage** | O(2^n × n) | Storing all 2^n subsets, avg size n |
| **Total** | O(n × 2^n) | Output dominates |

**Detailed Breakdown:**

1. **Recursion Depth (Call Stack):** O(n)
   - At any time, we have at most n recursive calls on stack
   - Longer chains: `include 1 → include 2 → include 3 → ... → include n`

2. **Output Result:** O(n × 2^n)
   - Total subsets = 2^n
   - Average elements per subset = n/2
   - Total storage = 2^n × (n/2) = O(n × 2^n)

3. **Working Space (Current Subset List):** O(n)
   - Reused across recursion (not counted separately)

**Final Space Complexity: O(n × 2^n)** (dominated by output)
- If we only count the recursion stack: O(n)
- If we count output: O(n × 2^n)

### Comparison with Other Approaches:

| Approach | Time | Space | Difficulty |
|----------|------|-------|-----------|
| **Recursive (This Solution)** | O(n·2^n) | O(n·2^n) | Medium |
| **Iterative** | O(n·2^n) | O(n·2^n) | Medium |
| **Bit Manipulation** | O(n·2^n) | O(n·2^n) | Hard |

## Key Takeaways

✅ **Backtracking Pattern:** Include → Recurse → Exclude → Recurse

✅ **Why `.copy()`:** Lists are mutable; without it, all results point to same object

✅ **Binary Decision Tree:** Each element = 1 choice, n elements = 2^n paths

✅ **Time = O(n·2^n):** We generate 2^n subsets, each taking O(n) to copy

✅ **Space = O(n·2^n):** Output storage + O(n) recursion stack

````python
from typing import List

class Solution:
    def subsets(self, nums: List[int]) -> List[List[int]]:
        """
        Generate all possible subsets of given array using recursion and backtracking.
        
        Args:
            nums: List of unique integers
            
        Returns:
            List of all possible subsets
        """
        result = []
        # Start recursion with index 0, empty subset, and result accumulator
        self.solve(nums, 0, [], result)
        return result

    def solve(self, nums, idx, subset, result):
        """
        Recursive helper function using backtracking.
        
        Args:
            nums: Original array
            idx: Current index being processed
            subset: Current subset being built
            result: Accumulator for all subsets
        """
        # BASE CASE: When we've considered all elements
        if idx >= len(nums):
            # Important: Use .copy() to store a deep copy, not a reference!
            # Without copy(), all results would reference the same list object
            result.append(subset.copy())
            return
        
        # ===== INCLUDE CURRENT ELEMENT =====
        # Add current element to subset
        subset.append(nums[idx])
        # Recurse with next element (current element is included)
        self.solve(nums, idx + 1, subset, result)
        
        # ===== EXCLUDE CURRENT ELEMENT (BACKTRACK) =====
        # Remove the element we just added (undo the choice)
        subset.pop()
        # Recurse with next element (current element is excluded)
        self.solve(nums, idx + 1, subset, result)

# Test Case 1: Example from problem
print("=" * 60)
print("Test Case 1: nums = [1,2,3]")
print("=" * 60)
nums = [1, 2, 3]
result = Solution().subsets(nums)
print(f"Output: {result}")
print(f"Number of subsets: {len(result)} (Expected: {2**len(nums)})")

# Test Case 2: Single element
print("\n" + "=" * 60)
print("Test Case 2: nums = [0] (Single element, including zero)")
print("=" * 60)
nums = [0]
result = Solution().subsets(nums)
print(f"Output: {result}")
print(f"Number of subsets: {len(result)} (Expected: {2**len(nums)})")

# Test Case 3: Negative numbers
print("\n" + "=" * 60)
print("Test Case 3: nums = [-1, -2] (Negative numbers)")
print("=" * 60)
nums = [-1, -2]
result = Solution().subsets(nums)
print(f"Output: {result}")
print(f"Number of subsets: {len(result)} (Expected: {2**len(nums)})")

# Test Case 4: Larger array
print("\n" + "=" * 60)
print("Test Case 4: nums = [1,2,3,4] (4 elements)")
print("=" * 60)
nums = [1, 2, 3, 4]
result = Solution().subsets(nums)
print(f"Output: {result}")
print(f"Number of subsets: {len(result)} (Expected: {2**len(nums)})")

# Test Case 5: Edge case with zero and negative
print("\n" + "=" * 60)
print("Test Case 5: nums = [-10, 0, 5] (Mixed: negative, zero, positive)")
print("=" * 60)
nums = [-10, 0, 5]
result = Solution().subsets(nums)
print(f"Output: {result}")
print(f"Number of subsets: {len(result)} (Expected: {2**len(nums)})")
````

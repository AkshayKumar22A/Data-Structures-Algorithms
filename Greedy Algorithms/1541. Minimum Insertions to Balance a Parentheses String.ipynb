```python
from collections import deque
class Solution:
    def minInsertions(self, s: str) -> int:
        stack = deque()
        for i in range(len(s)):
            if s[i] == ")":
                stack.append(s[i])
            if s[i] == "(":
                stack.append(s[i])
            if len(stack) >= 3 and stack[-1] == ")" and stack[-2] == ")" and stack[-3] == "(":
                stack.pop()
                stack.pop()
                stack.pop()
        print(stack)
        if not stack:
            return 0
        
        open = 0
        close = 0
        ans = 0
        while stack:
            temp = stack.pop()
            print(temp)
            if temp == ")":
                close += 1
                ans += 1
            if temp == ")" and close == 2:
                ans += 1
                close = 0
            if temp == "(":
                ans = ans + 2
        return ans
            

s = "(()))"
print(Solution().minInsertions(s))

s1 = "())"
print(Solution().minInsertions(s1))

s2 = "))())("
print(Solution().minInsertions(s2))
        
```

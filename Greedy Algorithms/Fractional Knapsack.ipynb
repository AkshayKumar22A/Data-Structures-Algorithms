# Fractional Knapsack
Given two arrays, val[] and wt[] , representing the values and weights of items, and an integer capacity representing the maximum weight a knapsack can hold, determine the maximum total value that can be achieved by putting items in the knapsack. You are allowed to break items into fractions if necessary.
Return the maximum value as a double, rounded to 6 decimal places.

Examples :

Input: val[] = [60, 100, 120], wt[] = [10, 20, 30], capacity = 50
Output: 240.000000
Explanation: By taking items of weight 10 and 20 kg and 2/3 fraction of 30 kg. Hence total price will be 60+100+(2/3)(120) = 240
Input: val[] = [500], wt[] = [30], capacity = 10
Output: 166.670000
Explanation: Since the item’s weight exceeds capacity, we take a fraction 10/30 of it, yielding value 166.670000.
Constraints:
1 ≤ val.size = wt.size ≤ 105
1 ≤ capacity ≤ 109
1 ≤ val[i], wt[i] ≤ 104

```python
class Solution:
    def fractionalKnapsack(self, val, wt, capacity):
        ind = list(range(len(val)))
        ind.sort(key = lambda i: val[i]/wt[i],reverse=True)
        val[:] = [val[i] for i in ind]
        wt[:] = [wt[i] for i in ind]

        currwt = 0
        currval = 0
        for i in range(len(val)):
            if currwt + wt[i] <= capacity:
                currwt += wt[i]
                currval += val[i]
            else:
                currval += ((val[i]//wt[i]) * (capacity - currwt))
                break
        return currval

val = [1, 5, 7, 2, 7, 10]
wt = [4, 9, 6, 3, 7, 3]
capacity = 24
print(Solution().fractionalKnapsack(val, wt, capacity))
```
